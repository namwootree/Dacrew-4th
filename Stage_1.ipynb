  - <font color=blue>**명목형**</font> : 값이 순서와 상관없이 의미만 구분되는 경우 (ex : 성별, 지역, 부서)<br><br>

  - <font color=blue>**이산형**</font> : 값을 하나하나 셀 수 있는 경우 (ex : 사기 거래 개수, 시간 당 결제 횟수)<br><br>
  - <font color=red>**연속형**</font> : 값을 하나하나 셀 수 없는 경우 (ex : 키, 온도)<br><br>

## 3 - 8 이상치 감지 (Anomaly Detection)

<br>

**신용카드 카드 사기 거래는 정상 거래와는 분명히 어딘가 이상할 것이라고 생각할 수 있습니다.**

예를 들어볼까요.

* 갑지기 큰 금액이 결제될 수도 있습니다.
* 서울에서 결제된 지 3분만에 부산에서 추가 결제될 수 도 있습니다.
* 신용카드 거래자의 신상이 불분명할 수 도 있습니다.<br><br>

이러한 사기 거래의 패턴의 경우 <font color=green>**보통의 정상 거래의 패턴과 다를 것**</font>이라는 것을 유추해볼 수 있습니다.

그렇다면 신용카드 거래에서의 데이터 또한 다른 데이터와 분명 이상한 패턴을 보일 것입니다.

이처럼 대부분의 보통 데이터와 다른 형태를 띄는 데이터를 <font color=red>**이상치** (*Anmoaly*)</font>라고 불립니다.<br><br>

이러한 이상치를 AI를 통해 식별하는 것을 <font color=blue>**이상치 감지** (*Anomaly Detection*)</font>라고 합니다.
