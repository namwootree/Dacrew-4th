  * Î≤îÏ£ºÍ∞Ä ÏûàÎäî Îç∞Ïù¥ÌÑ∞ Í∞íÏùÑ ÏßÅÏÇ¨Í∞ÅÌòïÏùò ÎßâÎåÄÎ°ú ÌëúÌòÑÌïòÎäî Í∑∏ÎûòÌîÑÏûÖÎãàÎã§.<br>

### Íµ¨ÏÑ± (Composition)

<br>

**<font color=red>ÌååÏù¥ Ï∞®Ìä∏ (Pie Chart)**</font><br><br>
  * Ï†ÑÏ≤¥Ïóê ÎåÄÌïú Í∞Å Î∂ÄÎ∂ÑÏùò ÎπÑÏú®ÏùÑ Î∂ÄÏ±ÑÍº¥ Î™®ÏñëÏúºÎ°ú Î∞±Î∂ÑÏú®Î°ú ÎÇòÌÉÄÎÇ¥Îäî Í∑∏ÎûòÌîÑÏûÖÎãàÎã§.<br>
  * Í∞Å Î∂ÄÏ±ÑÍº¥Ïùò Ï§ëÏã¨Í∞ÅÏù¥ Ï†ÑÏ≤¥ÏóêÏÑú Ï∞®ÏßÄÌïòÎäî ÎπÑÏú®ÏùÑ ÌïúÎààÏóê Î≥º Ïàò ÏûàÎã§Îäî Ïû•Ï†êÏù¥ ÏûàÏäµÎãàÎã§.<br><br>

````python
# V1 ÏπºÎüºÎßå ÏÇ¨Ïö©Ìï©ÎãàÎã§.
v1_train = train[['V1']]

# sklearn ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò cluster Ï§ëÏóêÏÑú KMeansÎ•º Î∂àÎü¨ÏòµÎãàÎã§.
# ÏûêÏÑ∏Ìïú ÎÇ¥Ïö©ÏùÄ Stage 2Î•º Ï∞∏Ï°∞Ìï¥Ï£ºÏÑ∏Ïöî.
from sklearn.cluster import KMeans

cluster = KMeans(n_clusters=3)

array_clusted = cluster.fit_predict(v1_train)

v1_clusted = pd.DataFrame(array_clusted)

# Íµ∞ÏßëÏóê Î™®Ïù∏ Îç∞Ïù¥ÌÑ∞Ïùò ÏàòÎ°ú Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ ÎßåÎì≠ÎãàÎã§.
v1_value_counts = pd.DataFrame(v1_clusted[0].value_counts())
v1_value_counts.columns = ['V1_Value_Counts']

display(v1_value_counts)
````

````python
x = v1_value_counts['V1_Value_Counts']
labels = v1_value_counts.index

plt.pie(
    x=x,# ÏÇ¨Ïö©Ìï† Îç∞Ïù¥ÌÑ∞
    labels=labels,# Îç∞Ïù¥ÌÑ∞Ïùò ÎùºÎ≤®
    startangle=70,# ÏãúÍ≥ÑÎ∞©Ìñ•ÏúºÎ°ú Í∞ÅÎèÑÎ•º ÎèåÎ¶¨ÎèÑÎ°ù ÏÑ§Ï†ï
    explode=[0.1, 0.1, 0.1], # Í∞ÅÍ∞ÅÏùò ÌååÏù¥Í∞Ä Îñ®Ïñ¥ÏßÄÎèÑÎ°ù ÏÑ§Ï†ï
    autopct = '%.1f%%' # ÌååÏù¥ Ï∞®Ìä∏ÏïàÏóê ÌçºÏÑºÌä∏ ÏàòÏπò ÏÇΩÏûÖ
    )

# Î≤îÎ°Ä ÏÉùÏÑ±
plt.legend(labels)

# XÏ∂ï Ïù¥Î¶ÑÏùÑ ÏÑ§Ï†ïÌï©ÎãàÎã§.
plt.xlabel('V1 Clustering')

# Í∑∏ÎûòÌîÑÏùÑ Ï†úÎ™©ÏùÑ ÏÑ§Ï†ïÌï©ÎãàÎã§.
plt.title('Pie Chart')

# ÏôÑÏÑ±Îêú Í∑∏ÎûòÌîÑÎ•º ÎèÑÏ∂úÌï©ÎãàÎã§.
plt.show()
````

<br>

<font color=red>**ÎàÑÏ†Å ÎßâÎåÄ Í∑∏ÎûòÌîÑ (Stacked Bar Chart)**</font><br><br>
  * Îç∞Ïù¥ÌÑ∞Ïùò Feature Îç∞Ïù¥ÌÑ∞Ïùò ÎπÑÏ∂©ÏùÑ ÎÇòÌÉÄÎÇ¥Îäî Í∑∏ÎûòÌîÑÏûÖÎãàÎã§.<br>
  * Îã§ÏñëÌïú Îç∞Ïù¥ÌÑ∞ FeatureÏùò Íµ¨ÏÑ± ÎπÑÏú®ÏùÑ ÌïúÎààÏóê Î≥º Ïàò ÏûàÎã§Îäî Ïû•Ï†êÏù¥ ÏûàÏäµÎãàÎã§.<br><br>

````python
# V1 ÏπºÎüºÎßå ÏÇ¨Ïö©Ìï©ÎãàÎã§.
v1_train = train[['V1']]
v2_train = train[['V2']]

# sklearn ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò cluster Ï§ëÏóêÏÑú KMeansÎ•º Î∂àÎü¨ÏòµÎãàÎã§.
# ÏûêÏÑ∏Ìïú ÎÇ¥Ïö©ÏùÄ Stage 2Î•º Ï∞∏Ï°∞Ìï¥Ï£ºÏÑ∏Ïöî.
from sklearn.cluster import KMeans

cluster = KMeans(n_clusters=3)

array_clusted_1 = cluster.fit_predict(v1_train)
array_clusted_2 = cluster.fit_predict(v1_train)

v1_clusted = pd.DataFrame(array_clusted_1)
v2_clusted = pd.DataFrame(array_clusted_2)

# Íµ∞ÏßëÏóê Î™®Ïù∏ Îç∞Ïù¥ÌÑ∞Ïùò ÏàòÎ°ú Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ ÎßåÎì≠ÎãàÎã§.
v1_value_counts = pd.DataFrame(v1_clusted[0].value_counts())
v1_value_counts.columns = ['V1_Value_Counts']

v2_value_counts = pd.DataFrame(v2_clusted[0].value_counts())
v2_value_counts.columns = ['V2_Value_Counts']


clusted_value_counts = pd.concat([v1_value_counts,
                                  v2_value_counts],
                                  axis=1)

display(clusted_value_counts.head())
````

````python
# Îπà ÎèÑÌôîÏßÄÎ•º ÏÉùÏÑ±Ìï©ÎãàÎã§.
# fig : Í∑∏ÎûòÌîÑÍ∞Ä Îã¥Í∏∞Í≤å ÎêòÎäî ÌîÑÎ†àÏûÑ (Ïï°Ïûê)
# ax : Ïã§Ï†ú Îç∞Ïù¥ÌÑ∞Í∞Ä ÎèÑÏãùÌôîÎêòÎäî Ï∫îÌçºÏä§ (ÎèÑÌôîÏßÄ)
fig, ax = plt.subplots()

# ÎèôÏùºÌïú ÎèÑÌôîÏßÄÏóê 'V1'Ïóê Í¥ÄÌïú ÎßâÎåÄ Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶∞Îã§.
ax.bar(clusted_value_counts.index,
        clusted_value_counts['V1_Value_Counts'],
        label='V1')

# ÎèôÏùºÌïú ÎèÑÌôîÏßÄÏóê '21'Ïóê Í¥ÄÌïú ÎßâÎåÄ Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶∞Îã§.
# ÏïÑÎûòÏóê 'V1' Í∑∏Î¶ºÏù¥ Îì§Ïñ¥Í∞ÄÎèÑÎ°ù ÌïúÎã§.
ax.bar(clusted_value_counts.index,
       clusted_value_counts['V2_Value_Counts'],
       bottom=clusted_value_counts['V1_Value_Counts'],
       label='V2')

# Î≤îÎ°Ä ÏÉùÏÑ±
plt.legend()

# XÏ∂ïÍ≥º YÏ∂ï Ïù¥Î¶ÑÏùÑ ÏÑ§Ï†ïÌï©ÎãàÎã§.
plt.xlabel('V1 & V2 Clustering')
plt.ylabel('V1 & V2 Value Count')

# Í∑∏ÎûòÌîÑÏùÑ Ï†úÎ™©ÏùÑ ÏÑ§Ï†ïÌï©ÎãàÎã§.
plt.title('Stacked Bar Chart')

# ÏôÑÏÑ±Îêú Í∑∏ÎûòÌîÑÎ•º ÎèÑÏ∂úÌï©ÎãàÎã§.
plt.show()
````
