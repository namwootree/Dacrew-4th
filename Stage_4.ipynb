### ÎπÑÍµê (Comparision)

<br>

**<font color=red>ÎßâÎåÄÏ∞®Ìä∏ (barplot)**</font><br><br>
  * Î≤îÏ£ºÍ∞Ä ÏûàÎäî Îç∞Ïù¥ÌÑ∞ Í∞íÏùÑ ÏßÅÏÇ¨Í∞ÅÌòïÏùò ÎßâÎåÄÎ°ú ÌëúÌòÑÌïòÎäî Í∑∏ÎûòÌîÑÏûÖÎãàÎã§.<br>
  * Í∞ÅÍ∞ÅÏùò Îç∞Ïù¥ÌÑ∞ ÏÖãÏùò Feature Î≥Ñ Îç∞Ïù¥ÌÑ∞Î•º ÎπÑÍµêÌïòÍ∏∞ Ï¢ãÏäµÎãàÎã§.<br><br>

```python
# Í∞ÅÍ∞ÅÏùò Feature Î≥Ñ ÏµúÎåÄÍ∞íÏùÑ Íµ¨Ìï©ÎãàÎã§.
max_train = pd.DataFrame(train.drop(columns=['ID']).max())
max_train.columns = ['MAX Value']

# Îç∞Ïù¥ÌÑ∞Ïùò ÏùºÎ∂ÄÎßå ÏÇ¨Ïö©Ìï©ÎãàÎã§.
max_train = max_train.iloc[:5]
display(max_train.head())
```

```python
# XÏ∂ïÍ≥º YÏ∂ïÏóê Îì§Ïñ¥Í∞à Í∞ÅÍ∞ÅÏùò Í∞íÏùÑ ÏßÄÏ†ïÌï©ÎãàÎã§.
x = list(max_train.index)
height = list(max_train['MAX Value'])

# ÎßâÎåÄ Ï∞®Ìä∏ ÏÉùÏÑ±
plt.bar(
    x=x, # XÏ∂ï
    height=height, # YÏ∂ï
    color=['b', 'g', 'b', 'g', 'r'] # ÎßâÎåÄÏùò ÏÉâÏÉÅÏùÑ ÏßÄÏ†ï
    )

# XÏ∂ïÍ≥º YÏ∂ï Ïù¥Î¶ÑÏùÑ ÏÑ§Ï†ïÌï©ÎãàÎã§.
plt.xlabel('Features')
plt.ylabel('Max Value')

# Í∑∏ÎûòÌîÑÏùÑ Ï†úÎ™©ÏùÑ ÏÑ§Ï†ïÌï©ÎãàÎã§.
plt.title('Bar Chart')

# ÏôÑÏÑ±Îêú Í∑∏ÎûòÌîÑÎ•º ÎèÑÏ∂úÌï©ÎãàÎã§.
plt.show()
```

<br>

**<font color=red>ÏÑ† Ï∞®Ìä∏ (lineplot)**</font><br><br>
  * Ïó∞ÏÜçÌòï Îç∞Ïù¥ÌÑ∞ Í∞íÏùÑ ÌïòÎÇòÏùò ÏÑ†ÏúºÎ°ú ÌëúÌòÑÌïòÎäî Í∑∏ÎûòÌîÑÏûÖÎãàÎã§.<br>
  * ÏãúÍ∞Ñ Í≤ΩÍ≥ºÏóê Îî∞Î•∏ Îç∞Ïù¥ÌÑ∞ Î≥ÄÌôîÎ•º ÎπÑÍµêÌïòÍ∏∞ Ï¢ãÏäµÎãàÎã§.<br><br>

```python
# 'V1' ÏπºÎüºÎßå ÏÇ¨Ïö©Ìï©ÎãàÎã§.
v1_train = train[['V1']]

# Îç∞Ïù¥ÌÑ∞ ÏùºÎ∂ÄÎßå ÏÇ¨Ïö©Ìï©ÎãàÎã§.
v1_train = v1_train.iloc[:7]

display(v1_train.head())
```

```python
# XÏ∂ïÍ≥º YÏ∂ïÏóê Îì§Ïñ¥Í∞à Í∞ÅÍ∞ÅÏùò Í∞íÏùÑ ÏßÄÏ†ïÌï©ÎãàÎã§.
x = list(v1_train.index)
Y = list(v1_train['V1'])

# ÎßâÎåÄ Ï∞®Ìä∏ ÏÉùÏÑ±
plt.plot(
    x, # XÏ∂ï
    Y, # YÏ∂ï
    c='r', # ÏÑ† ÏÉâÏÉÅ ÏÑ§Ï†ï
    marker='o', # ÎßàÏª§ Î™®Ïñë ÏÑ§Ï†ï
    mfc='b' # ÎßàÏª§ ÏÉâÏÉÅ ÏÑ§Ï†ï 
    )

# XÏ∂ïÍ≥º YÏ∂ï Ïù¥Î¶ÑÏùÑ ÏÑ§Ï†ïÌï©ÎãàÎã§.
plt.xlabel('V1 Feature')
plt.ylabel('V1 Value')

# Í∑∏ÎûòÌîÑÏùÑ Ï†úÎ™©ÏùÑ ÏÑ§Ï†ïÌï©ÎãàÎã§.
plt.title('Line Chart')

# ÏôÑÏÑ±Îêú Í∑∏ÎûòÌîÑÎ•º ÎèÑÏ∂úÌï©ÎãàÎã§.
plt.show()
```

### Î∂ÑÌè¨ (Distribution)

<br>

**<font color=red>ÌûàÏä§ÌÜ†Í∑∏Îû® (Histogram)**</font><br><br>
  * Ïó∞ÏÜçÌòï Îç∞Ïù¥ÌÑ∞ Í∞íÏùò Î∂ÑÌè¨Î•º ÎßâÎåÄÎ°ú ÌëúÌòÑÌïòÎäî Í∑∏ÎûòÌîÑÏûÖÎãàÎã§.<br>
  * ÌäπÏ†ï Îç∞Ïù¥ÌÑ∞ ÏÖãÏùò FeatureÏùò Îç∞Ïù¥ÌÑ∞ Î∂ÑÌè¨Î•º ÌôïÏù∏ÌïòÍ∏∞ Ï¢ãÏäµÎãàÎã§.<br><br>

```python
# 'V1' ÏπºÎüºÎßå ÏÇ¨Ïö©Ìï©ÎãàÎã§.
v1_train = train[['V1']]

display(v1_train.head())
```

```python
x = v1_train['V1']

plt.hist(
    x=x, # XÏ∂ï
    bins=100, # Íµ¨Í∞Ñ Í∞úÏàò ÏÑ§Ï†ï
    color='green' # ÏÉâÏÉÅ ÏÑ§Ï†ï
    )

plt.xlim([-15, 3]) # XÏ∂ï Î≤îÏúÑ
plt.ylim([0, 20000]) # yÏ∂ï Î≤îÏúÑ

# XÏ∂ïÍ≥º YÏ∂ï Ïù¥Î¶ÑÏùÑ ÏÑ§Ï†ïÌï©ÎãàÎã§.
plt.xlabel('V1 Value')
plt.ylabel('V1 Value Count')

# Í∑∏ÎûòÌîÑÏùÑ Ï†úÎ™©ÏùÑ ÏÑ§Ï†ïÌï©ÎãàÎã§.
plt.title('Histogram')

# ÏôÑÏÑ±Îêú Í∑∏ÎûòÌîÑÎ•º ÎèÑÏ∂úÌï©ÎãàÎã§.
plt.show()
```

**<font color=red>ÌååÏù¥ Ï∞®Ìä∏ (Pie Chart)**</font><br><br>
  * Ï†ÑÏ≤¥Ïóê ÎåÄÌïú Í∞Å Î∂ÄÎ∂ÑÏùò ÎπÑÏú®ÏùÑ Î∂ÄÏ±ÑÍº¥ Î™®ÏñëÏúºÎ°ú Î∞±Î∂ÑÏú®Î°ú ÎÇòÌÉÄÎÇ¥Îäî Í∑∏ÎûòÌîÑÏûÖÎãàÎã§.<br>
  * Í∞Å Î∂ÄÏ±ÑÍº¥Ïùò Ï§ëÏã¨Í∞ÅÏù¥ Ï†ÑÏ≤¥ÏóêÏÑú Ï∞®ÏßÄÌïòÎäî ÎπÑÏú®ÏùÑ ÌïúÎààÏóê Î≥º Ïàò ÏûàÎã§Îäî Ïû•Ï†êÏù¥ ÏûàÏäµÎãàÎã§.<br><br>

```python
# V1 ÏπºÎüºÎßå ÏÇ¨Ïö©Ìï©ÎãàÎã§.
v1_train = train[['V1']]

# sklearn ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò cluster Ï§ëÏóêÏÑú KMeansÎ•º Î∂àÎü¨ÏòµÎãàÎã§.
# ÏûêÏÑ∏Ìïú ÎÇ¥Ïö©ÏùÄ Stage 2Î•º Ï∞∏Ï°∞Ìï¥Ï£ºÏÑ∏Ïöî.
from sklearn.cluster import KMeans

cluster = KMeans(n_clusters=3)

array_clusted = cluster.fit_predict(v1_train)

v1_clusted = pd.DataFrame(array_clusted)

# Íµ∞ÏßëÏóê Î™®Ïù∏ Îç∞Ïù¥ÌÑ∞Ïùò ÏàòÎ°ú Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ ÎßåÎì≠ÎãàÎã§.
v1_value_counts = pd.DataFrame(v1_clusted[0].value_counts())
v1_value_counts.columns = ['V1_Value_Counts']

display(v1_value_counts)
```

```python
x = v1_value_counts['V1_Value_Counts']
labels = v1_value_counts.index

plt.pie(
    x=x,# ÏÇ¨Ïö©Ìï† Îç∞Ïù¥ÌÑ∞
    labels=labels,# Îç∞Ïù¥ÌÑ∞Ïùò ÎùºÎ≤®
    startangle=70,# ÏãúÍ≥ÑÎ∞©Ìñ•ÏúºÎ°ú Í∞ÅÎèÑÎ•º ÎèåÎ¶¨ÎèÑÎ°ù ÏÑ§Ï†ï
    explode=[0.1, 0.1, 0.1], # Í∞ÅÍ∞ÅÏùò ÌååÏù¥Í∞Ä Îñ®Ïñ¥ÏßÄÎèÑÎ°ù ÏÑ§Ï†ï
    autopct = '%.1f%%' # ÌååÏù¥ Ï∞®Ìä∏ÏïàÏóê ÌçºÏÑºÌä∏ ÏàòÏπò ÏÇΩÏûÖ
    )

# Î≤îÎ°Ä ÏÉùÏÑ±
plt.legend(labels)

# XÏ∂ï Ïù¥Î¶ÑÏùÑ ÏÑ§Ï†ïÌï©ÎãàÎã§.
plt.xlabel('V1 Clustering')

# Í∑∏ÎûòÌîÑÏùÑ Ï†úÎ™©ÏùÑ ÏÑ§Ï†ïÌï©ÎãàÎã§.
plt.title('Pie Chart')

# ÏôÑÏÑ±Îêú Í∑∏ÎûòÌîÑÎ•º ÎèÑÏ∂úÌï©ÎãàÎã§.
plt.show()
```
