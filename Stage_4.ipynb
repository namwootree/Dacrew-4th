## 3 - 1 데이터 시각화란 무엇인가요?

<br>

**데이터 시각화**

  * 지금 시대는 다양한 정보로 넘쳐나고 있습니다. 하지만 중요한 것과 그렇지 않은 것을 이해하기 위해 정보를 살펴 보는 것이 점점 더 어려워지고 있습니다.

  * 이럴 때 **데이터 분석 과정에서 더 좋은 인사이트를 발견하기 위해서** 혹은 **생성된 모든 정보로 데이터를 이해하기** 위하여 적합한 형태의 데이터로 시각화하는 것입니다.

<br>

<br>

**데이터 시각화의 필요성**<br><br>

  * <font color=darkblue>**많은 양의 데이터를 한눈에 볼 수 있습니다.**</font> 모니터에 띄워진 엑셀 화면 영역을 벗어날 만큼 많은 양의 데이터를 상상해보세요. 어떻게 데이터를 보고, 데이터의 변화를 파악할 수 있을까요?<br><br>

  * <font color=darkblue>**데이터 분석에 대한 전문 지식이 없어도, 누구나 쉽게 데이터 인사이트를 찾을 수 있습니다.**</font> 차트의 시각적 패턴을 근거로 데이터의 추세는 어떤지, 다른 값들과 구별되는 이상값을 빠르게 찾을 수 있습니다. <br><br>

  * <font color=darkblue>**요약 통계보다 정확한 데이터 분석 결과를 도출할 수 있습니다.** </font> Anscomes’ Quarter는 동일한 요약 통계(평균, 표준편차, 상관관계)를 가진 4개의 데이터 셋(set)을 산점도로 시각화하였을 때, 시각적 패턴이 명확히 다르다는 것을 입증합니다.<br><br>

  * <font color=darkblue>**효과적인 데이터 인사이트 공유로 데이터 기반의 의사결정을 할 수 있습니다.**</font> 시각화 자료는 사람들의 머릿속에 빠르게 인지되고 기억될 뿐만 아니라, 메시지의 스토리텔링에 대한 공감을 더 합니다.<br><br>

### 비교 (Comparision)

<br>

**<font color=red>막대차트 (barplot)**</font><br><br>
  * 범주가 있는 데이터 값을 직사각형의 막대로 표현하는 그래프입니다.<br>
  * 각각의 데이터 셋의 Feature 별 데이터를 비교하기 좋습니다.<br><br>

```python
# 각각의 Feature 별 최대값을 구합니다.
max_train = pd.DataFrame(train.drop(columns=['ID']).max())
max_train.columns = ['MAX Value']

# 데이터의 일부만 사용합니다.
max_train = max_train.iloc[:5]
display(max_train.head())
```

```python
# X축과 Y축에 들어갈 각각의 값을 지정합니다.
x = list(max_train.index)
height = list(max_train['MAX Value'])

# 막대 차트 생성
plt.bar(
    x=x, # X축
    height=height, # Y축
    color=['b', 'g', 'b', 'g', 'r'] # 막대의 색상을 지정
    )

# X축과 Y축 이름을 설정합니다.
plt.xlabel('Features')
plt.ylabel('Max Value')

# 그래프을 제목을 설정합니다.
plt.title('Bar Chart')

# 완성된 그래프를 도출합니다.
plt.show()
```

<br>

**<font color=red>선 차트 (lineplot)**</font><br><br>
  * 연속형 데이터 값을 하나의 선으로 표현하는 그래프입니다.<br>
  * 시간 경과에 따른 데이터 변화를 비교하기 좋습니다.<br><br>

```python
# 'V1' 칼럼만 사용합니다.
v1_train = train[['V1']]

# 데이터 일부만 사용합니다.
v1_train = v1_train.iloc[:7]

display(v1_train.head())
```

```python
# X축과 Y축에 들어갈 각각의 값을 지정합니다.
x = list(v1_train.index)
Y = list(v1_train['V1'])

# 막대 차트 생성
plt.plot(
    x, # X축
    Y, # Y축
    c='r', # 선 색상 설정
    marker='o', # 마커 모양 설정
    mfc='b' # 마커 색상 설정 
    )

# X축과 Y축 이름을 설정합니다.
plt.xlabel('V1 Feature')
plt.ylabel('V1 Value')

# 그래프을 제목을 설정합니다.
plt.title('Line Chart')

# 완성된 그래프를 도출합니다.
plt.show()
```

### 구성 (Composition)

<br>

**<font color=red>파이 차트 (Pie Chart)**</font><br><br>
  * 전체에 대한 각 부분의 비율을 부채꼴 모양으로 백분율로 나타내는 그래프입니다.<br>
  * 각 부채꼴의 중심각이 전체에서 차지하는 비율을 한눈에 볼 수 있다는 장점이 있습니다.<br><br>

```python
# V1 칼럼만 사용합니다.
v1_train = train[['V1']]

# sklearn 라이브러리의 cluster 중에서 KMeans를 불러옵니다.
# 자세한 내용은 Stage 2를 참조해주세요.
from sklearn.cluster import KMeans

cluster = KMeans(n_clusters=3)

array_clusted = cluster.fit_predict(v1_train)

v1_clusted = pd.DataFrame(array_clusted)

# 군집에 모인 데이터의 수로 데이터프레임을 만듭니다.
v1_value_counts = pd.DataFrame(v1_clusted[0].value_counts())
v1_value_counts.columns = ['V1_Value_Counts']

display(v1_value_counts)
```

```python
x = v1_value_counts['V1_Value_Counts']
labels = v1_value_counts.index

plt.pie(
    x=x,# 사용할 데이터
    labels=labels,# 데이터의 라벨
    startangle=70,# 시계방향으로 각도를 돌리도록 설정
    explode=[0.1, 0.1, 0.1], # 각각의 파이가 떨어지도록 설정
    autopct = '%.1f%%' # 파이 차트안에 퍼센트 수치 삽입
    )

# 범례 생성
plt.legend(labels)

# X축 이름을 설정합니다.
plt.xlabel('V1 Clustering')

# 그래프을 제목을 설정합니다.
plt.title('Pie Chart')

# 완성된 그래프를 도출합니다.
plt.show()
```
