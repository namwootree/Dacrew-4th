### ÌÖåÏù¥Î∏î Ìï∏Îì§ÎßÅÏùÑ ÌôúÏö©Ìïú Ïù¥ÏÉÅÏπò Î≥ÄÌôò

**Ïù¥ÏÉÅÏπòÎ•º Íµ¨Î∂ÑÌïòÎäî Í∏∞Ï§Ä IQR**

  * IQR (Interquartile range)Ïù¥ÎûÄ Ïù¥ÏÉÅÏπòÎ•º Íµ¨Î∂ÑÌï† Ïàò ÏûàÎäî Î∞©Î≤ïÎ°† Ï§ë ÌïòÎÇòÏûÖÎãàÎã§.

  * ÏÇ¨Î∂ÑÏúÑÏàòÎ•º ÌôúÏö©ÌôîÏó¨ Ïù¥ÏÉÅÏπòÎ•º ÏãùÎ≥ÑÌïòÎäî Í∏∞Ï§ÄÏÑ† Îëê Í∞ú (ÏµúÎåÄÍ∞íÍ≥º ÏµúÏÜåÍ∞í)Î•º ÏÑ§Ï†ïÌï©ÎãàÎã§.

  * ÏµúÎåÄÍ∞íÍ≥º ÏµúÏÜåÍ∞íÏùÑ Î≤óÏñ¥ÎÇú Îç∞Ïù¥ÌÑ∞Î•º Ïù¥ÏÉÅÏπòÎùºÍ≥† Í∑úÏ†ïÌï©ÎãàÎã§.

  * ÏµúÎåÄÍ∞íÍ≥º ÏµúÏÜåÍ∞íÏùÄ Îã§ÏùåÏùò ÏàòÏãùÏúºÎ°ú Íµ¨Ìï† Ïàò ÏûàÏäµÎãàÎã§.

    - IQR = 3ÏÇ¨Î∂ÑÏúÑÏàò - 1ÏÇ¨Î∂ÄÏúÑÏàò (ÏÉÅÏúÑ 75% ÏßÄÏ†ê Í∞íÍ≥º ÌïòÏúÑ 25% ÏßÄÏ†ê Í∞íÏùò Ï∞®Ïù¥)

    - ÏµúÎåÄÍ∞í = 3ÏÇ¨Î∂ÑÏúÑÏàò + 1.5 * IQR

    - ÏµúÏÜåÍ∞í = 1ÏÇ¨Î∂ÑÏúÑÏàò - 1.5 * IQR

**ÏÇ¨Î∂ÑÏúÑÏàò Íµ¨Ìï¥Ï£ºÎäî Ìï®Ïàò quantile()**



```
ÏÉÅÏúÑ 75% ÏßÄÏ†ê Í∞í (3ÏÇ¨Î∂ÑÏúÑÏàò) = Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.75)

ÌïòÏúÑ 25% ÏßÄÏ†ê Í∞í (1ÏÇ¨Î∂ÑÏúÑÏàò) = Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.25)
```



**IQR Í∏∞Ï§ÄÏóê Ìï¥ÎãπÌïòÎäî Ïù¥ÏÉÅÏπòÎ•º Îã§Î•∏ Í∞íÏúºÎ°ú Î≥ÄÍ≤ΩÌïòÎäî Î∞©Î≤ï**

```
# IQRÎ•º ÌôúÏö©Ìïú Ïù¥ÏÉÅÏπò Í∏∞Ï§Ä ÏÑ§Ï†ï
IQR = Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.75) - Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.25)
maximum = Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.75) + (1.5 * IQR)
minimum = Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.25) - (1.5 * IQR)

ÌäπÏ†ï Ï°∞Í±¥ 1 = (Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'] > maximum)
ÌäπÏ†ï Ï°∞Í±¥ 2 = (Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'] < minimum)

# Ïù¥ÏÉÅÏπò Í∏∞Ï§ÄÏóê Î∂ÄÌï©Ìïú Îç∞Ïù¥ÌÑ∞Î•º Îã§Î•∏ Í∞íÏúºÎ°ú ÎåÄÏ≤¥

## maximumÏóê Ï¥àÍ≥ºÌïú Îç∞Ïù¥ÌÑ∞Î•º 3ÏÇ¨Î∂ÑÏúÑÍ∞íÏúºÎ°ú Î≥ÄÍ≤Ω
Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ.loc[ÌäπÏ†ï Ï°∞Í±¥ 1] = Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.75)

## minimumÏóê ÎØ∏ÎßåÏù∏ Îç∞Ïù¥ÌÑ∞Î•º 1ÏÇ¨Î∂ÑÏúÑÍ∞íÏúºÎ°ú Î≥ÄÍ≤Ω
Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ.loc[ÌäπÏ†ï Ï°∞Í±¥ 2] = Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.25)
```



IQRÎ•º Í∏∞Ï§ÄÏúºÎ°ú Ïù¥ÏÉÅÏπò ÌòπÏùÄ Í∑πÎã®Í∞íÏù¥ Îã§Î•∏ Í∞íÏúºÎ°ú Î≥ÄÍ≤ΩÎêòÏóàÍ∏∞Ïóê 'V1' ÏπºÎüºÏùò ÌëúÏ§ÄÌé∏Ï∞®Í∞Ä Í∞êÏÜåÌïòÏòÄÏùåÏùÑ ÌôïÏù∏Ìï† Ïàò ÏûàÏäµÎãàÎã§.

````python
# 'V1' ÏπºÎüºÏóê IQRÎ•º Í∏∞Ï§ÄÏúºÎ°ú Ïù¥ÏÉÅÏπò Îç∞Ïù¥ÌÑ∞ Î≥ÄÌôò

print(f'Í∏∞Ï°¥ V1 ÏπºÎüº ÌëúÏ§ÄÌé∏Ï∞® : {train["V1"].std()}\n')

IQR = train['V1'].quantile(0.75) - train['V1'].quantile(0.25)
maximum = train['V1'].quantile(0.75) + (1.5 * IQR)
minimum = train['V1'].quantile(0.25) - (1.5 * IQR)

print(f'IQR : {IQR}')
print(f'maximum : {maximum}')
print(f'minimum : {minimum}\n')

cond1 = (train['V1'] > maximum)
cond2 = (train['V1'] < minimum)

train.loc[cond1] = train['V1'].quantile(0.75)
train.loc[cond2] = train['V1'].quantile(0.25)

print(f'Ïù¥ÏÉÅÏπò Ï≤òÎ¶¨Îêú V1 ÏπºÎüº ÌëúÏ§ÄÌé∏Ï∞® : {train["V1"].std()}')
````

ÏÉòÌîåÎßÅ Í∏∞Î≤ïÏóêÎäî ÌÅ¨Í≤å Ïñ∏Îçî ÏÉòÌîåÎßÅÍ≥º Ïò§Î≤Ñ ÏÉòÌîåÎßÅÏù¥ ÏûàÏäµÎãàÎã§.

  * Ïñ∏Îçî ÏÉòÌîåÎßÅ (Under Sampling)

    - Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ ÏàòÏóê ÎßûÍ≤å Ï§ÑÏù¥Îäî Î∞©ÏãùÏûÖÎãàÎã§.

  * Ïò§Î≤Ñ ÏÉòÌîåÎßÅ (Over Sampling)
    
    - ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ ÏàòÏóê ÎßûÍ≤å ÎäòÎ¶¨Îäî Î∞©ÏãùÏûÖÎãàÎã§.

    - ÎèôÏùºÌïú (ÌòπÏùÄ Ïú†ÏÇ¨Ìïú) ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º ÌôúÏö©ÌïòÍ∏∞Ïóê ÏÜåÏàò Î≤îÏ£ºÏùò ÌäπÏ†ï Ï†ïÎ≥¥Í∞Ä Í≥ºÏûâ Î∞òÏòÅÎê† Ïàò ÏûàÏäµÎãàÎã§.

````python
# Î∂àÍ∑†Ìòï Îç∞Ïù¥ÌÑ∞ Î¨∏Ï†úÎ•º ÌôïÏù∏Ìïú 'Val' Îç∞Ïù¥ÌÑ∞ ÏÖãÏùÑ ÌôúÏö©Ìï©ÎãàÎã§. 

val = pd.read_csv('./val.csv')

print(val['Class'].value_counts())
````

````python
# 'val' Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ Î≥µÏÇ¨ÌïòÏó¨ ÎèôÏùºÌïú Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ ÏÇ¨Ïö©Ìï©ÎãàÎã§.
df = val.copy()

# Î∂àÍ∑†Ìòï Îç∞Ïù¥ÌÑ∞ Î¨∏Ï†úÎ•º Ìï¥Í≤∞Ìï† ÏπºÎüºÏùÑ Íµ¨Î∂ÑÌï©ÎãàÎã§.
# y : Î∂àÍ∑†Ìòï Îç∞Ïù¥ÌÑ∞ Î¨∏Ï†úÍ∞Ä Î∞úÏÉùÌïú ÌäπÏÑ± / X : y Ïù¥Ïô∏Ïùò ÏπºÎüºÎì§ 
y = df[['Class']]
X = df.drop(columns=['Class'])
````

**Random Under Sampler**



Î¨¥ÏûëÏúÑÎ°ú Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º Ï†úÍ±∞ÌïòÏó¨ Î∂àÍ∑†Ìòï Îç∞Ïù¥ÌÑ∞ Î¨∏Ï†úÎ•º Ìï¥Í≤∞Ìï©ÎãàÎã§.

````python
# imblearn ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò under_sampling Ï§ëÏóêÏÑú RandomUnderSamplerÎ•º Î∂àÎü¨ÏòµÎãàÎã§.
from imblearn.under_sampling import RandomUnderSampler
````

````python
# ÏÉòÌîåÎßÅÏùÑ Ïö©Ïù¥ÌïòÍ≤å ÏÇ¨Ïö©ÌïòÍ≤å Î≥ÑÎèÑÎ°ú Î≥ÄÏàòÎ°ú ÏßÄÏ†ïÌï©ÎãàÎã§.
under_sampler = RandomUnderSampler()

under_sampled_X, under_sampled_y = under_sampler.fit_resample(X, y)
````

````python
# Pandas ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò concat Ìï®ÏàòÎ•º ÌôúÏö©ÌïòÏó¨ Î∂ÑÎ¶¨Îêú XÏôÄ yÎ•º Îã§Ïãú Ìï©Ï≥êÏ§çÎãàÎã§.
df_random_under = pd.concat([under_sampled_X, under_sampled_y], axis=1)
````

````python
# Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ ÏàòÏôÄ ÎèôÏùºÌïòÍ≤å Î¨¥ÏûëÏúÑÎ°ú Ï∂îÏ∂úÌïòÏòÄÍ∏∞Ïóê Ï¥ù Îç∞Ïù¥ÌÑ∞Ïùò ÏàòÍ∞Ä Í∞êÏÜåÌïòÏòÄÏäµÎãàÎã§.
print(df_random_under.shape)
````

**Condensed Nearest Neighbour**

Condensed Nearest NeighbourÎäî Îã§Ïùå Í≥ºÏ†ïÏùÑ Í±∞Ï≥ê Under SamplingÏùÑ ÏàòÌñâÌï©ÎãàÎã§.

  1. ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ Î™®Îëê ÏßëÌï© SÏóê Ìè¨Ìï®ÌïúÎã§.

  2. Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ Ï§ëÏóêÏÑú ÌïòÎÇòÎ•º Í≥®ÎùºÏÑú Í∞ÄÏû• Í∞ÄÍπåÏö¥ Îç∞Ïù¥ÌÑ∞Í∞Ä Îã§Ïàò Î≤îÏ£º Îç∞Ïù¥ÌÑ∞Í∞Ä ÏïÑÎãàÎ©¥ ÏßëÌï© SÏóê Ìè¨Ìï®ÌïúÎã§.

  3. Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Îçî Ïù¥ÏÉÅ ÏÑ†ÌÉùÏßÄÍ∞Ä ÏóÜÏùÑ ÎïåÍπåÏßÄ Í≥ºÏ†ï 2Î•º Î∞òÎ≥µÌïúÎã§.

````python
# imblearn ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò under_sampling Ï§ëÏóêÏÑú CondensedNearestNeighbourÎ•º Î∂àÎü¨ÏòµÎãàÎã§.
from imblearn.under_sampling import CondensedNearestNeighbour
````

````python
# ÏÉòÌîåÎßÅÏùÑ Ïö©Ïù¥ÌïòÍ≤å ÏÇ¨Ïö©ÌïòÍ≤å Î≥ÑÎèÑÎ°ú Î≥ÄÏàòÎ°ú ÏßÄÏ†ïÌï©ÎãàÎã§.
under_sampler = CondensedNearestNeighbour()

under_sampled_X, under_sampled_y = under_sampler.fit_resample(X, y)
````

````python
# Pandas ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò concat Ìï®ÏàòÎ•º ÌôúÏö©ÌïòÏó¨ Î∂ÑÎ¶¨Îêú XÏôÄ yÎ•º Îã§Ïãú Ìï©Ï≥êÏ§çÎãàÎã§.
df_cnn_under = pd.concat([under_sampled_X, under_sampled_y], axis=1)
````

````python
# Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ ÏàòÏôÄ ÎèôÏùºÌïòÍ≤å Î¨¥ÏûëÏúÑÎ°ú Ï∂îÏ∂úÌïòÏòÄÍ∏∞Ïóê Ï¥ù Îç∞Ïù¥ÌÑ∞Ïùò ÏàòÍ∞Ä Í∞êÏÜåÌïòÏòÄÏäµÎãàÎã§.
print(df_cnn_under.shape)
````

````python
# ÏïåÍ≥†Î¶¨Ï¶ò ÌäπÏÑ±ÏÉÅ Ï∂îÏ∂úÎêú Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ ÏàòÍ∞Ä ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ ÏàòÏôÄ Í∞ôÏßÄ ÏïäÏùÑ Ïàò ÏûàÏäµÎãàÎã§.
print(df_cnn_under['Class'].value_counts())
````

**SMOTE**

SMOTE (Synthetic Minority Over-sampling Technique)Îäî Îã§ÏùåÏùò Í≥ºÏ†ïÏùÑ ÏßÑÌñâÌï©ÎãàÎã§.

  1. ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ Ï§ëÏóêÏÑú Î¨¥ÏûëÏúÑÎ°ú ÌïòÎÇòÎ•º ÏÑ†ÌÉùÌï©ÎãàÎã§.
  
  2. KNN (K-nearest neighbors) ÏïåÍ≥†Î¶¨Ï¶òÏùÑ ÌôúÏö©ÌïòÏó¨ Í∞ÄÏû• Í∞ÄÍπåÏö¥ KÍ∞úÏùò ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º Ï∞æÏäµÎãàÎã§.

  3. Ï∞æÏùÄ KÍ∞úÏùò ÏÜåÏàò Î≤îÏ£º Îç∞Ïù¥ÌÑ∞ Ï§ë ÌïòÎÇòÎ•º Î¨¥ÏûëÏúÑÎ°ú ÏÑ†ÌÉùÌï©ÎãàÎã§.

  4. Ï≤òÏùåÏóê ÏÑ†ÌÉùÌïú Îç∞Ïù¥ÌÑ∞ÏôÄ KNNÎ•º ÌÜµÌï¥ ÏÑ†ÌÉùÎêú Îç∞Ïù¥ÌÑ∞ ÏÇ¨Ïù¥Ïóê ÏÑ†ÏùÑ Í∏ãÏäµÎãàÎã§.

  5. Í∑∏ ÏÑ† ÏúÑÏóê ÏûàÎäî ÏûÑÏùòÏùò Ï†êÏùÑ ÏÑ†ÌÉùÌïòÏó¨ ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞Î•º ÏÉùÏÑ±Ìï©ÎãàÎã§.

  6. ÏõêÌïòÎäî ÎßåÌÅºÏùò Îç∞Ïù¥ÌÑ∞ ÏàòÍ∞Ä ÏÉùÏÑ±Îê† ÎïåÍπåÏßÄ ÏúÑÏùò Í≥ºÏ†ïÏùÑ Î∞òÎ≥µÌï©ÎãàÎã§.

````python
# imblearn ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò over_sampling Ï§ëÏóêÏÑú SMOTEÎ•º Î∂àÎü¨ÏòµÎãàÎã§.
from imblearn.over_sampling import SMOTE
````

````python
# ÏÉòÌîåÎßÅÏùÑ Ïö©Ïù¥ÌïòÍ≤å ÏÇ¨Ïö©ÌïòÍ≤å Î≥ÑÎèÑÎ°ú Î≥ÄÏàòÎ°ú ÏßÄÏ†ïÌï©ÎãàÎã§.
over_sampler = SMOTE()

over_sampled_X, over_sampled_y = over_sampler.fit_resample(X, y)
````

````python
# Pandas ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò concat Ìï®ÏàòÎ•º ÌôúÏö©ÌïòÏó¨ Î∂ÑÎ¶¨Îêú XÏôÄ yÎ•º Îã§Ïãú Ìï©Ï≥êÏ§çÎãàÎã§.
df_smote_over = pd.concat([over_sampled_X, over_sampled_y], axis=1)
````

Îã§Î•∏ Î≥ÄÏàòÎì§ÍπåÏßÄ Ï¢ÖÌï©Ï†ÅÏúºÎ°ú Í≥†Î†§ÌïòÏó¨ Ïú†ÏÇ¨Ìïú ÏÑ±Í≤© (ÎπÑÏä∑Ìïú Ìå®ÌÑ¥Í≥º Íµ¨Ï°∞)ÏùÑ Í∞ÄÏßÑ Îç∞Ïù¥ÌÑ∞Î•º Í∞ôÏùÄ Íµ∞ÏßëÏóê Î¨∂Ïñ¥ Íµ¨Î∂ÑÌï† Ïàò ÏûàÏäµÎãàÎã§.

Ïù¥Í≤ÉÏùÑ ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ (Clustering) Ïù¥ÎùºÍ≥† Ìï©ÎãàÎã§.

ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅÏù¥ ÏûòÎêú Í≤ΩÏö∞, ÌäπÏ†ï Ìïú Íµ∞ÏßëÏóêÎäî Îç∞Ïù¥ÌÑ∞Í∞ÑÏóê ÎÜíÏùÄ Ïú†ÏÇ¨ÏÑ±ÏùÑ Î≥¥Ïù¥Í≥† Íµ∞ÏßëÍ∞ÑÏóêÎäî ÎÇÆÏùÄ Ïú†ÏÇ¨ÏÑ±ÏùÑ Î≥¥Ïó¨Ï§çÎãàÎã§.

**K-Means**

K-Means ÏïåÍ≥†Î¶¨Ï¶òÏùÄ Íµ∞ÏßëÏùò Ï§ëÏã¨Í≥º Îç∞Ïù¥ÌÑ∞Í∞ÑÏùò ÌèâÍ∑† Í±∞Î¶¨Î•º ÏÇ¨Ïö©ÌïòÏó¨ KÍ∞úÏùò Íµ∞ÏßëÏùÑ ÎßåÎì§Ïñ¥ ÎÇ¥Îäî Í≤ÉÏûÖÎãàÎã§.

K-Means Îäî Îã§ÏùåÏùò Í≥ºÏ†ïÏùÑ ÏßÑÌñâÌï©ÎãàÎã§.


  1. KÍ∞úÏùò ÏûÑÏùò Ï§ëÏã¨Ï†êÏùÑ Î∞∞ÏπòÌï©ÎãàÎã§.

  2. Í∞ÅÍ∞ÅÏùò Îç∞Ïù¥ÌÑ∞Îäî Ï§ëÏã¨Ï†êÍ≥ºÏùò ÌèâÍ∑†Í±∞Î¶¨Î•º Í∏∞Ï§ÄÏúºÎ°ú Í∞ÄÏû• Í∞ÄÍπåÏö¥ Ï§ëÏã¨Ï†êÏùÑ Ï§ëÏã¨ÏúºÎ°ú Íµ∞ÏßëÏùÑ ÌòïÏÑ±Ìï©ÎãàÎã§.

  3. Ï§ëÏã¨Ï†êÏùÑ Íµ∞ÏßëÏùò Ï§ëÏã¨ÏúºÎ°ú Ïù¥ÎèôÌï©ÎãàÎã§.

  4. Í≤∞Í≥ºÍ∞Ä Î∞îÎÄåÏßÄ ÏïäÍ±∞ÎÇò 'max_iter' ÌïòÏù¥Ìçº ÌååÎùºÎØ∏ÌÑ∞Ïóê ÏÑ§Ï†ïÎêú Í∞íÍπåÏßÄ 2Î≤àÍ≥º 3Î≤à Í≥ºÏ†ïÏùÑ Î∞òÎ≥µÌï©ÎãàÎã§.

````python
val = pd.read_csv('./val.csv')

# 'ID'ÏôÄ 'Class' Î≥ÄÏàòÎäî Ï†úÏô∏Ìï©ÎãàÎã§.
df = val.drop(columns=['ID', 'Class'])

# 'Class' Î≥ÄÏàòÎßå Í∞ÄÏ†∏ÏòµÎãàÎã§.
label = val[['Class']]
````

````python
# sklearn ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò cluster Ï§ëÏóêÏÑú KMeansÎ•º Î∂àÎü¨ÏòµÎãàÎã§.
from sklearn.cluster import KMeans
````

````python
# ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅÏùÑ Ïö©Ïù¥ÌïòÍ≤å ÏÇ¨Ïö©ÌïòÍ≤å Î≥ÑÎèÑÎ°ú Î≥ÄÏàòÎ°ú ÏßÄÏ†ïÌï©ÎãàÎã§.
# 'n_clusters' ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞Î•º ÏÇ¨Ïö©Ìï¥ ÎßåÎì§Ïñ¥ÎÇº Íµ∞ÏßëÏùò Í∞úÏàòÎ•º ÏÑ§Ï†ïÌï©ÎãàÎã§.
cluster = KMeans(n_clusters=2)

array_clusted = cluster.fit_predict(df)

df_clusted = pd.DataFrame(array_clusted)
````

````python
# Pandas ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò concat Ìï®ÏàòÎ•º ÌôúÏö©ÌïòÏó¨ ÌÅ¥Î†ÅÏä§ÌÑ∞ÎßÅ Í≤∞Í≥ºÏôÄ 'Class'Î•º Îã§Ïãú Ìï©Ï≥êÏ§çÎãàÎã§.
df_result = pd.concat([df_clusted, label], axis=1)

# ÏπºÎüº Î™ÖÏùÑ Î≥ÄÍ≤ΩÌï¥Ï§çÎãàÎã§

df_result.columns = ['Í≤∞Í≥º', 'Ï†ïÎãµ']
````

````python
# ÎåÄÎã§ÏàòÏùò Îç∞Ïù¥ÌÑ∞Îäî Ïûò Íµ¨Î∂ÑÌïòÏòÄÏßÄÎßå Î™á Í∞ú Îç∞Ïù¥ÌÑ∞Îäî ÌãÄÎ¶¥ Í≤ÉÏùÑ ÌôïÏù∏Ìï† Ïàò ÏûàÏäµÎãàÎã§.
print(df_result['Í≤∞Í≥º'].value_counts())
print(df_result['Ï†ïÎãµ'].value_counts())
````

**Mean-Shift Clustering**

Mean-Shift Clustering ÏïåÍ≥†Î¶¨Ï¶òÏùÄ Îç∞Ïù¥ÌÑ∞Í∞Ä Î™®Ïó¨ÏûàÎäî Î∞ÄÎèÑÎ•º Í∏∞Ï§ÄÏúºÎ°ú Íµ∞ÏßëÏùÑ ÌòïÏÑ±Ìï©ÎãàÎã§.

K-MeansÏôÄ ÎòêÎã§Î•∏ Ï∞®Ïù¥Ï†êÏù¥ ÏûàÎã§Î©¥ K-MeansÎäî Ï¥àÍ∏∞Ïóê Íµ∞ÏßëÏùò Í∞úÏàòÎ•º ÏÑ§Ï†ïÌï¥Ï£ºÏñ¥Ïïº ÌïòÏßÄÎßå Mean-ShiftÎäî ÏÑ§Ï†ïÌï¥Ï§Ä ÎåÄÏó≠Ìè≠Ïùò Î≤îÏúÑÏóê Îî∞Îùº ÏûêÎèôÏ†ÅÏúºÎ°ú Íµ∞ÏßëÏùò Í∞úÏàòÍ∞Ä Ï†ïÌï¥ÏßëÎãàÎã§.

````python
val = pd.read_csv('./val.csv')

# 'ID'ÏôÄ 'Class' Î≥ÄÏàòÎäî Ï†úÏô∏Ìï©ÎãàÎã§.
df = val.drop(columns=['ID', 'Class'])

# 'Class' Î≥ÄÏàòÎßå Í∞ÄÏ†∏ÏòµÎãàÎã§.
label = val[['Class']]
````

````python
# sklearn ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò cluster Ï§ëÏóêÏÑú MeanShiftÎ•º Î∂àÎü¨ÏòµÎãàÎã§.
from sklearn.cluster import MeanShift
````

````python
# ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅÏùÑ Ïö©Ïù¥ÌïòÍ≤å ÏÇ¨Ïö©ÌïòÍ≤å Î≥ÑÎèÑÎ°ú Î≥ÄÏàòÎ°ú ÏßÄÏ†ïÌï©ÎãàÎã§.
# 'bandwidth' ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞Î•º ÏÇ¨Ïö©Ìï¥ ÎåÄÏó≠Ìè≠Ïùò Î≤îÏúÑÎ•º ÏÑ§Ï†ïÌï©ÎãàÎã§.
cluster = MeanShift(bandwidth=2)

array_clusted = cluster.fit_predict(df)

df_clusted = pd.DataFrame(array_clusted)
````

````python
# Pandas ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò concat Ìï®ÏàòÎ•º ÌôúÏö©ÌïòÏó¨ ÌÅ¥Î†ÅÏä§ÌÑ∞ÎßÅ Í≤∞Í≥ºÏôÄ 'Class'Î•º Îã§Ïãú Ìï©Ï≥êÏ§çÎãàÎã§.
df_result = pd.concat([df_clusted, label], axis=1)

# ÏπºÎüº Î™ÖÏùÑ Î≥ÄÍ≤ΩÌï¥Ï§çÎãàÎã§

df_result.columns = ['Í≤∞Í≥º', 'Ï†ïÎãµ']
````
