## 3 - 1 전처리와 특성 공학은 무엇인가요?

<br>

**전처리 (Preprocessing)**

  * 우리가 시험을 준비하는 모습을 떠올려 봅시다. 교재에 오타가 있으면 수정하기도 합니다. 그리고 교재에서 설명하는 내용이 어려우면 이해하기 쉬운 형태로 바꾸기도 합니다.

  * **데이터 분석 과정에서 더 좋은 인사이트를 발견하기 위해서** 혹은 **머신러닝 학습 과정에서 더 나은 성능을 내기 위해**, <font color=red>**기존의 데이터를 잘 정돈되게 정리**</font>하여 위의 과정에 적합한 형태의 데이터로 가공하는 것입니다.

<br>

<br>

**특성 공학 (Feature Engineering)**

  * 다시 한 번 더 시험을 준비하는 모습을 떠올려 봅시다. 공부하는 과목의 핵심 개념을 이해하기위해 좋은 정리 노트를 찾기도 하고 만들기도 합니다. 그리고 시험 직전에는 시험 출제 빈도가 높은 개념만 따로 적은 노트만 보는 친구들의 모습을 보기도 합니다.

  * 특징 공학은 전처리와 마찬가지로 머신러닝 학습 과정에서 더 나은 성능을 내기 위해, <font color=red>**데이터의 특징이 잘 드러내도록**</font> 새로운 특징을 만들어내거나 불필요한 특징은 제거하는 방법 등으로 데이터를 가공하는 것을 뜻합니다.

<br>

<br>

**전처리와 특성 공학의 필요성**<br><br>

  * <font color=darkblue>**결측치를 제거하거나 다른 값으로 채워야 합니다.**</font> Stage 1에서 배웠 듯이 데이터 수집 과정이 완벽하지 않아 데이터 값이 존재하지 않은 결측치가 발생할 수 있습니다. 결측치가 존재할 수 있습니다.<br><br>

  * <font color=darkblue>**특정 데이터 수가 부족할 수 도 있습니다.**</font> 기존의 데이터가 우리에게 인사이트를 주기에는 불완전할 수 있습니다. 예를 들어 보면 '축구를 좋아하는 고객의 남녀 각각의 특징 조사'에서 여성 고객에 관하여 인사이트를 얻기에는 여성 고객 데이터 수가 부족할 수 있습니다.<br><br>

  * <font color=darkblue>**예외적인 상황을 고려할 필요가 있습니다.**</font> 예를 들어 '치킨에 대하여 소비액에 조사하였을 때, 월드컵 기간의 치킨 소비량이 일반적인 치킨 소비량'이라고 보기는 힘듭니다.<br><br>

  * <font color=darkblue>**도메인 지식을 통해 좋은 데이터 특징을 발굴할 수 도 있습니다.**</font> 신용카드 결재에 미치는 중요한 요소로 불경기 여부, 축제 혹은 명절 여부 등이 있습니다.<br><br>

  * <font color=darkblue>**반대로 불필요한 데이터 특징은 제거할 수 도 있습니다.**</font> MZ세대는 너무 광범위한 나이대를 포함합니다. 그래서 MZ세대 여부를 물어보지 않는 것이 고객의 인사이트를 찾아내기 더 쉬울 수 있습니다.<br><br>

# 원하는 조건 작성

**원하는 조건에 다양할 수 있습니다.** 

**각각의 경우를 살펴봅시다.**

<br>

**<font color=red>교집합</font> : 특정 조건 1에 부합하고 특정 조건 2에도 부합한 데이터**



```
특정 조건 1 = (데이터프레임['특정 칼럼 1'] >= 특정 값)
특정 조건 2 = (데이터프레임['특정 칼럼 2'] <= 특정 값)

데이터프레임.loc[(특정 조건 1 & 특정 조건 2)]
```

<br>


````python
# V1 칼럼에서 -0.1보다 크면서 V2에서 0.1보다 작은 데이터만 추출
cond1 = (train['V1'] >= -0.1)
cond2 = (train['V2'] <= 0.1)

display(train.loc[cond1 & cond2])
````

<br>

**<font color=red>합집합</font> : 특정 조건 1에 부합하기도 하고 특정 조건 2에도 부합하기도 한 데이터**



```
특정 조건 1 = (데이터프레임['특정 칼럼 1'] >= 특정 값)
특정 조건 2 = (데이터프레임['특정 칼럼 2'] <= 특정 값)

데이터프레임.loc[(특정 조건 1 | 특정 조건 2)]
```

<br>


````python
# V1 칼럼에서 -0.1보다 크거나 V2에서 0.1보다 작은 데이터만 추출
cond1 = (train['V1'] >= -0.1)
cond2 = (train['V2'] <= 0.1)

display(train.loc[cond1 | cond2])
````

<br>

**<font color=red>여집합</font> : 특정 조건에 부합하지 않은 데이터**



```
특정 조건 = (데이터프레임['특정 칼럼 1'] >= 특정 값)

데이터프레임.loc[(~특정 조건)]
```

<br>


````python
# 'V1' 칼럼에서 0 이하가 아닌 데이터만 추출

cond = (train['V1'] <= 0)

display(train.loc[~cond])
````

### 결측치 처리

<br>

결측치를 처리할 수 있는 방법 크게 결측치가 있는 데이터를 삭제하는 <font color=green>**제거** (*Delection*)</font>와 결측치를 다른 값으로 채우는 <font color=blue>**대치** (*Imputation*)</font>가 있습니다.

<br>

우리가 사용하는 데이터 셋에는 결측치가 존재하지 않습니다.

그래서 **인위적으로 결측치를 만들어 보겠습니다.**

<br>

````python
# random 라이브러리를 활용하여 무작위 수를 생성할 수 있습니다.
# randrange는 무작위 정수 값을 생성할 수 있습니다.
from random import randrange

# numpy 라이버러리를 활용하여 array 데이터를 다룰 수 있습니다.
# NaN는 결측값을 생성합니다.
from numpy import NaN
````

````python
# 'train' 데이터에서 'V1' 칼럼만 사용합니다
df = train[['V1']]

# 무작위로 'df' 데이터의 값을 결측값으로 변경합니다
list_null = []
for _ in range(100):

  index_null = randrange(df.shape[0])
  list_null.append(index_null)

list_null = sorted(list_null)

df.iloc[list_null] = NaN
````

````python
# 결측치로 변경된 데이터의 인덱스
print(list_null[:5])
````

````python
# Stage 1를 기억하며 결측값이 몇 개 존재하는 지 확인해봅시다
df.isnull().sum()
````

<br><br>

**제거**

<br>

````python
# 결측치가 존재하는 데이터를 제거해봅시다.

new_df = df.dropna()

# 결측치가 모두 제거되었음을 확인했습니다.
print(new_df.isnull().sum())
````

````python
# 그러나 우리가 활용할 수 있는 데이터의 수가 줄어들었음을 확인할 수 있습니다.
print(df.shape[0], new_df.shape[0])
````

<br><br>

**대치**

<br>

````python
# 결측치를 다른 값으로 변경해봅시다.

# 결측치를 특정 수로 대치합니다.
df_0 = df.fillna(0)
display(df_0.iloc[[1357]])

# 결측치를 평균으로 대치합니다.
df_mean = df.fillna(df['V1'].mean())
display(df_mean.iloc[[1357]])

# 결측치를 바로 이전 값으로으로 대치합니다.
df_pad = df.fillna(method='pad')
display(df_pad.iloc[1355:1359])

# 결측치를 바로 이후 값으로으로 대치합니다.
df_bfill = df.fillna(method='bfill')
display(df_bfill.iloc[1355:1359])
````

**이상치를 구분하는 기준 IQR**<br><br>

  * <font color=red>**IQR** (*Interquartile range*)</font>이란 이상치를 구분할 수 있는 방법론 중 하나입니다.<br><br>

  * <font color=green>**사분위수를 활용**</font>하여 이상치를 식별하는 기준선 두 개 (최대값과 최소값)를 설정합니다.<br><br>

  * 최대값과 최소값을 <font color=blue>**벗어난 데이터를 이상치라고 규정**</font>합니다.<br><br>

  * 최대값과 최소값은 다음의 수식으로 구할 수 있습니다.<br><br>

    - IQR = 3사분위수 - 1사부위수 (상위 75% 지점 값과 하위 25% 지점 값의 차이)

    - 최대값 = 3사분위수 + 1.5 * IQR

    - 최소값 = 1사분위수 - 1.5 * IQR

<br>

<br>

**사분위수 구해주는 함수 <font color=red>quantile()</font>**



```
상위 75% 지점 값 (3사분위수) = 데이터프레임['특정 칼럼'].quantile(0.75)

하위 25% 지점 값 (1사분위수) = 데이터프레임['특정 칼럼'].quantile(0.25)
```

<br>


<br>

**IQR 기준에 해당하는 이상치를 다른 값으로 변경하는 방법**

```
# IQR를 활용한 이상치 기준 설정
IQR = 데이터프레임['특정 칼럼'].quantile(0.75) - 데이터프레임['특정 칼럼'].quantile(0.25)
maximum = 데이터프레임['특정 칼럼'].quantile(0.75) + (1.5 * IQR)
minimum = 데이터프레임['특정 칼럼'].quantile(0.25) - (1.5 * IQR)

특정 조건 1 = (데이터프레임['특정 칼럼'] > maximum)
특정 조건 2 = (데이터프레임['특정 칼럼'] < minimum)

# 이상치 기준에 부합한 데이터를 다른 값으로 대체

## maximum에 초과한 데이터를 3사분위값으로 변경
데이터프레임.loc[특정 조건 1] = 데이터프레임['특정 칼럼'].quantile(0.75)

## minimum에 미만인 데이터를 1사분위값으로 변경
데이터프레임.loc[특정 조건 2] = 데이터프레임['특정 칼럼'].quantile(0.25)
```

<br>

<br>

IQR를 기준으로 이상치 혹은 극단값이 다른 값으로 변경되었기에 'V1' 칼럼의 **표준편차가 감소**하였음을 확인할 수 있습니다.

<br>

````python
# 'V1' 칼럼에 IQR를 기준으로 이상치 데이터 변환

print(f'기존 V1 칼럼 표준편차 : {train["V1"].std()}\n')

IQR = train['V1'].quantile(0.75) - train['V1'].quantile(0.25)
maximum = train['V1'].quantile(0.75) + (1.5 * IQR)
minimum = train['V1'].quantile(0.25) - (1.5 * IQR)

print(f'IQR : {IQR}')
print(f'maximum : {maximum}')
print(f'minimum : {minimum}\n')

cond1 = (train['V1'] > maximum)
cond2 = (train['V1'] < minimum)

train.loc[cond1] = train['V1'].quantile(0.75)
train.loc[cond2] = train['V1'].quantile(0.25)

print(f'이상치 처리된 V1 칼럼 표준편차 : {train["V1"].std()}')
````

### 스케일링

<br>

여러분들은 IQR를 활용하여 데이터의 Feature의 범위를 조정시켜 보았습니다.

이러한 과정을 <font color=red>**데이터 스케일링** (*Data Scaling*)</font>이라고 합니다.

데이터 스케일링에는 다양한 기법들이 존재합니다.

하나씩 알아가보도록 합시다.

#### Standard Scaler

<br>

Standard Scaler는 데이터의 Feature의 <font color=darkblue>**평균을 0**</font> 그리고 <font color=darkblue>**분산을 1**</font>로 스케일링하여 데이터 <font color=red>**표준 정규 분포**</font>를 따르도록 합니다.

<br>

stand.png

````python
# 'train' 데이터 셋에서 'ID' 칼럼은 제외하고 사용합니다.
df = train.drop(columns=['ID'])
````

````python
# sklearn 라이브러리에서 preprocessing 중에서 StandardScaler를 불러옵니다.
from sklearn.preprocessing import StandardScaler
````

````python
# 스케일링을 용이하게 사용하게 별도로 변수로 지정합니다.
stand_scaler = StandardScaler()

# fit_transform 함수를 활용해 StandardScaler를 적용시킨 후 array 형태로 반환합니다.
array_stand_scaled = stand_scaler.fit_transform(df)

# array 형태의 데이터를 pandas 라이브러리의 DataFrame 함수를 사용해 데이터프레임로 변환합니다.
df_stand_scaled = pd.DataFrame(array_stand_scaled)

display(df_stand_scaled.head())
````

#### 언더 샘플링 VS 오버 샘플링

<br>

불균형 데이터 문제를 샘플링을 통해 해결할 수 있습니다.

샘플링이란 불균형한 데이터 분포를 균등하게 맞추기 위해 특정 데이터를 추출해가는 방식입니다.

샘플링 기법에는 크게 언더 샘플링과 오버 샘플링이 있습니다.<br><br>

  * <font color=red>**언더 샘플링** (*Under Sampling*)</font><br><br>

    - <font color=darkblue>**다수 범주의 데이터를 소수 범주의 데이터 수에 맞게 줄이는 방식**</font>입니다.<br><br>

    - 다수 범주의 데이터를 모두 활용하지 않기에 다수 범주의 데이터에서 <font color=#1589FF>**유의미한 정보를 손실**</font>할 수 있습니다.<br><br>


  * <font color=red>**오버 샘플링** (*Over Sampling*)</font><br><br>
    
    - <font color=darkblue>**소수 범주의 데이터를 다수 범주의 데이터 수에 맞게 늘리는 방식**</font>입니다.<br><br>

    - 동일한 (혹은 유사한) 소수 범주의 데이터를 활용하기에 <font color=#1589FF>**소수 범주의 특정 정보가 과잉 반영**</font>될 수 있습니다.<br><br>

````python
# 불균형 데이터 문제를 확인한 'Val' 데이터 셋을 활용합니다. 

val = pd.read_csv('./val.csv')

print(val['Class'].value_counts())
````

````python
# 'val' 데이터프레임을 복사하여 동일한 데이터프레임을 사용합니다.
df = val.copy()

# 불균형 데이터 문제를 해결할 칼럼을 구분합니다.
# y : 불균형 데이터 문제가 발생한 특성 / X : y 이외의 칼럼들 
y = df[['Class']]
X = df.drop(columns=['Class'])
````

#### 언더 샘플링

<br>

<font color=red>**Random Under Sampler**</font>



<br>

<font color=darkblue>**무작위로 다수 범주의 데이터를 제거**</font>하여 불균형 데이터 문제를 해결합니다.

<br>

````python
# imblearn 라이브러리의 under_sampling 중에서 RandomUnderSampler를 불러옵니다.
from imblearn.under_sampling import RandomUnderSampler
````

````python
# 샘플링을 용이하게 사용하게 별도로 변수로 지정합니다.
under_sampler = RandomUnderSampler()

under_sampled_X, under_sampled_y = under_sampler.fit_resample(X, y)
````

````python
# Pandas 라이브러리의 concat 함수를 활용하여 분리된 X와 y를 다시 합쳐줍니다.
df_random_under = pd.concat([under_sampled_X, under_sampled_y], axis=1)
````

````python
# 다수 범주의 데이터를 소수 범주의 데이터 수와 동일하게 무작위로 추출하였기에 총 데이터의 수가 감소하였습니다.
print(df_random_under.shape)
````

````python
# 다수 범주의 데이터 (정상 거래)의 수를 소수 범주의 거래의 데이터 (사기 거래)의 수와 동일하게 맞추어져습니다.
print(df_random_under['Class'].value_counts())
````

<font color=red>**Near Miss**</font>

<br>

Near Miss는 <font color=darkblue>**소수 범주의 데이터와 근처에 있는 다수 범주의 데이터를 추출**</font>합니다.

그리고 '근처에 있다'라는 말의 정의하는 3가지 방법에 따라 각기 다른 방식으로 다수 범주의 데이터를 추출합니다.

NearMiss에서 제공하는 'version' 하이퍼 파라미터 (우리가 설정할 수 있는 값)에 따라 3가지 다른 Under Sampling 방법론을 제시합니다.

<br>

````python
# imblearn 라이브러리의 under_sampling 중에서 NearMiss를 불러옵니다.
from imblearn.under_sampling import NearMiss
````

<br>

* **Version 1**<br><br>

  - 특정 다수 범주의 데이터로부터 가장 가까운 N개 (n_neighbors)의 소수 범주의 데이터까지의 평균 거리를 구합니다.<br><br>

  - 계산된 평균 거리가 가장 가까운 M개의 데이터는 남기고 나머지는 제거합니다.<br><br>

* **Version 2**<br><br>

  - 특정 다수 범주의 데이터로부터 가장 먼 거리에 있는 N개 (n_neighbors)의 소수 클래스 샘플까지의 평균 거리를 구합니다.<br><br>

  - 계산된 평균 거리가 가장 가까운 M개의 데이터는 남기고 나머지는 제거합니다.<br><br>

* **Version 3**<br><br>

  - 모든 소수 범주의 데이터로부터 M개의 가장 가까운 다수 범주의 데이터를 선택한다.<br><br>

  - 선택된 다수 범주의 데이터에서부터 가장 가까운 N개 (n_neighbors)의 소수 범주의 데이터까지의 거리를 계산한다.<br><br>

  - 평균 거리가 가장 먼 다수 범주의 데이터를 남기고 나머지는 제거합니다.<br><br>

````python
# 샘플링을 용이하게 사용하게 별도로 변수로 지정합니다.
# NearMiss의 다양한 version을 사용하고 싶으면 해당 version 외의 것은 주석처리하면 됩니다.
under_sampler = NearMiss(version=1, n_neighbors=3)
# under_sampler = NearMiss(version=2, n_neighbors=3)
# under_sampler = NearMiss(version=3, n_neighbors=3)

under_sampled_X, under_sampled_y = under_sampler.fit_resample(X, y)
````

````python
# Pandas 라이브러리의 concat 함수를 활용하여 분리된 X와 y를 다시 합쳐줍니다.
df_near_under = pd.concat([under_sampled_X, under_sampled_y], axis=1)
````

````python
# 다수 범주의 데이터를 소수 범주의 데이터 수와 동일하게 무작위로 추출하였기에 총 데이터의 수가 감소하였습니다.
print(df_near_under.shape)
````

````python
# 다수 범주의 데이터 (정상 거래)의 수를 소수 범주의 거래의 데이터 (사기 거래)의 수와 동일하게 맞추어져습니다.
print(df_near_under['Class'].value_counts())
````

<font color=red>**Condensed Nearest Neighbour**</font>

<br>

Condensed Nearest Neighbour는 다음 과정을 거쳐 Under Sampling을 수행합니다.<br><br>

  1. 소수 범주의 데이터 모두 집합 S에 포함한다.<br><br>

  2. 다수 범주의 데이터 중에서 하나를 골라서 <font color=darkblue>**가장 가까운 데이터가 다수 범주 데이터가 아니면 집합 S에 포함**</font>한다.<br><br>

  3. 다수 범주의 데이터에서 더 이상 선택지가 없을 때까지 과정 2를 반복한다.<br><br>

````python
# imblearn 라이브러리의 under_sampling 중에서 CondensedNearestNeighbour를 불러옵니다.
from imblearn.under_sampling import CondensedNearestNeighbour
````

````python
# 샘플링을 용이하게 사용하게 별도로 변수로 지정합니다.
under_sampler = CondensedNearestNeighbour()

under_sampled_X, under_sampled_y = under_sampler.fit_resample(X, y)
````

````python
# Pandas 라이브러리의 concat 함수를 활용하여 분리된 X와 y를 다시 합쳐줍니다.
df_cnn_under = pd.concat([under_sampled_X, under_sampled_y], axis=1)
````

````python
# 다수 범주의 데이터를 소수 범주의 데이터 수와 동일하게 무작위로 추출하였기에 총 데이터의 수가 감소하였습니다.
print(df_cnn_under.shape)
````

````python
# 알고리즘 특성상 추출된 다수 범주의 데이터 수가 소수 범주의 데이터 수와 같지 않을 수 있습니다.
print(df_cnn_under['Class'].value_counts())
````

<font color=red>**SMOTE**</font>

<br>

SMOTE (Synthetic Minority Over-sampling Technique)는 다음의 과정을 진행합니다.<br><br>

  1. 소수 범주의 데이터 중에서 무작위로 하나를 선택합니다.<br><br>
  
  2. <font color=darkblue>**KNN (K-nearest neighbors) 알고리즘을 활용**</font>하여 가장 가까운 K개의 소수 범주의 데이터를 찾습니다.<br><br>

  3. 찾은 K개의 소수 범주 데이터 중 하나를 무작위로 선택합니다.<br><br>

  4. 처음에 선택한 데이터와 KNN를 통해 선택된 데이터 사이에 선을 긋습니다.<br><br>

  5. <font color=darkblue>**그 선 위에 있는 임의의 점을 선택하여 새로운 데이터를 생성**</font>합니다.<br><br>

  6. 원하는 만큼의 데이터 수가 생성될 때까지 위의 과정을 반복합니다.<br><br>

````python
# imblearn 라이브러리의 over_sampling 중에서 SMOTE를 불러옵니다.
from imblearn.over_sampling import SMOTE
````

````python
# 샘플링을 용이하게 사용하게 별도로 변수로 지정합니다.
over_sampler = SMOTE()

over_sampled_X, over_sampled_y = over_sampler.fit_resample(X, y)
````

````python
# Pandas 라이브러리의 concat 함수를 활용하여 분리된 X와 y를 다시 합쳐줍니다.
df_smote_over = pd.concat([over_sampled_X, over_sampled_y], axis=1)
````

````python
# 소수 범주의 데이터를 다수 범주의 데이터 수와 동일하게 생성하였기에 총 데이터의 수가 증가하였습니다.
print(df_smote_over.shape)
````

````python
# 소수 범주의 데이터 (사기 거래)의 수를 다수 범주의 거래의 데이터 (정상 거래)의 수와 동일하게 맞추어져습니다.
print(df_random_over['Class'].value_counts())
````

````python
# 동일한 데이터가 몇 개가 있는 지 duplicated 함수를 사용하여 알아봅시다.
# 기존 데이터를 바탕으로 새롭게 데이터를 생성하였기에 상대적으로 중복된 데이터가 적습니다.
df_same = df_random_over.duplicated()
print(df_same.sum())
````

<font color=red>**ADASYN**</font>

<br>

ADASYN (Adaptive Synthetic Sampling)는 기존의 SMOTE를 보완한 기법입니다.

SMOTE 알고리즘에서 소수 범주의 데이터 주위에 다수 범주의 데이터가 얼마나 있는 지 고려합니다.

결론적으로 소수 범주의 데이터 주변에 다수 범주의 데이터가 많을 경우 해당 소수 범주 데이터와 유사한 데이터를 더 많이 생성합니다.

즉, <font color=darkblue>**소수 범주의 데이터 중에서 다수 범주의 데이터와 가까이 있는 데이터를 중심으로 유사한 데이터를 더 생성**</font>합니다.

<br>

````python
# imblearn 라이브러리의 over_sampling 중에서 ADASYN를 불러옵니다.
from imblearn.over_sampling import ADASYN
````

````python
# 샘플링을 용이하게 사용하게 별도로 변수로 지정합니다.
over_sampler = ADASYN()

over_sampled_X, over_sampled_y = over_sampler.fit_resample(X, y)
````

````python
# Pandas 라이브러리의 concat 함수를 활용하여 분리된 X와 y를 다시 합쳐줍니다.
df_adasyn_over = pd.concat([over_sampled_X, over_sampled_y], axis=1)
````

````python
# 소수 범주의 데이터를 다수 범주의 데이터 수와 동일하게 생성하였기에 총 데이터의 수가 증가하였습니다.
print(df_adasyn_over.shape)
````

````python
# 소수 범주의 데이터 (사기 거래)의 수를 다수 범주의 거래의 데이터 (정상 거래)의 수가 유사하게 맞추어져습니다.
# 동일하지 않는 이유는 생성하고자 하는 데이터의 수가 소수 범주의 데이터 주변의 다수 범주의 데이터 비율에 영향을 받기 떄문입니다. 
print(df_adasyn_over['Class'].value_counts())
````

````python
# 동일한 데이터가 몇 개가 있는 지 duplicated 함수를 사용하여 알아봅시다.
# 기존 데이터를 바탕으로 새롭게 데이터를 생성하였기에 상대적으로 중복된 데이터가 적습니다.
df_same = df_adasyn_over.duplicated()
print(df_same.sum())
````

### 클러스터링

<br>

우리는 파생변수 파트에서 하나의 변수에서 데이터의 범위에 따라 데이터를 구분지어 보았습니다.

이처럼 <font color=darkblue>**유사한 성격 (비슷한 패턴과 구조)을 가진 데이터를 같은 군집으로 묶어 구분**</font>할 수 있습니다.

이것을 <font color=red>**클러스터링** (*Clustering*)</font> 이라고 합니다.

<br>

클러스터링이 잘된 경우, 특정 한 군집에는 데이터간에 높은 유사성을 보이고 군집간에는 낮은 유사성을 보여줍니다.

<br>

<font color=red>**K-Means**</font>

<br>

K-Means 알고리즘은 <font color=darkblue>**군집의 중심과 데이터간의 평균 거리를 사용**</font>하여 K개의 군집을 만들어 내는 것입니다.<br><br>

K-Means 는 다음의 과정을 진행합니다.<br><br>


  1. K개의 임의 중심점을 배치합니다.<br><br>

  2. 각각의 데이터는 중심점과의 평균거리를 기준으로 가장 가까운 중심점을 중심으로 군집을 형성합니다.<br><br>

  3. 중심점을 군집의 중심으로 이동합니다.<br><br>

  4. 결과가 바뀌지 않거나 'max_iter' 하이퍼 파라미터에 설정된 값까지 2번과 3번 과정을 반복합니다.<br><br>

````python
val = pd.read_csv('./val.csv')

# 'ID'와 'Class' 변수는 제외합니다.
df = val.drop(columns=['ID', 'Class'])

# 'Class' 변수만 가져옵니다.
label = val[['Class']]
````

````python
# sklearn 라이브러리의 cluster 중에서 KMeans를 불러옵니다.
from sklearn.cluster import KMeans
````

````python
# 클러스터링을 용이하게 사용하게 별도로 변수로 지정합니다.
# 'n_clusters' 하이퍼파라미터를 사용해 만들어낼 군집의 개수를 설정합니다.
cluster = KMeans(n_clusters=2)

array_clusted = cluster.fit_predict(df)

df_clusted = pd.DataFrame(array_clusted)
````

````python
# Pandas 라이브러리의 concat 함수를 활용하여 클렁스터링 결과와 'Class'를 다시 합쳐줍니다.
df_result = pd.concat([df_clusted, label], axis=1)

# 칼럼 명을 변경해줍니다

df_result.columns = ['결과', '정답']
````

````python
# 대다수의 데이터는 잘 구분하였지만 몇 개 데이터는 틀릴 것을 확인할 수 있습니다.
print(df_result['결과'].value_counts())
print(df_result['정답'].value_counts())
````

<font color=red>**Mean-Shift Clustering**</font>

<br>

Mean-Shift Clustering 알고리즘은 <font color=darkblue>**데이터가 모여있는 밀도를 기준으로 군집을 형성**</font>합니다.

K-Means와 또다른 차이점이 있다면 K-Means는 초기에 군집의 개수를 설정해주어야 하지만 Mean-Shift는 설정해준 대역폭의 범위에 따라 자동적으로 군집의 개수가 정해집니다.

<br>

````python
val = pd.read_csv('./val.csv')

# 'ID'와 'Class' 변수는 제외합니다.
df = val.drop(columns=['ID', 'Class'])

# 'Class' 변수만 가져옵니다.
label = val[['Class']]
````

````python
# sklearn 라이브러리의 cluster 중에서 MeanShift를 불러옵니다.
from sklearn.cluster import MeanShift
````

````python
# 클러스터링을 용이하게 사용하게 별도로 변수로 지정합니다.
# 'bandwidth' 하이퍼파라미터를 사용해 대역폭의 범위를 설정합니다.
cluster = MeanShift(bandwidth=0.1)

array_clusted = cluster.fit_predict(df)

df_clusted = pd.DataFrame(array_clusted)
````

````python
# Pandas 라이브러리의 concat 함수를 활용하여 클렁스터링 결과와 'Class'를 다시 합쳐줍니다.
df_result = pd.concat([df_clusted, label], axis=1)

# 칼럼 명을 변경해줍니다

df_result.columns = ['결과', '정답']
````

````python
# Mean-Shift가 대역폭에 따라 2개의 군집이 아닌 다수의 군집을 생성함을 확인할 수 있습니다.
print(df_result['결과'].value_counts())
print(df_result['정답'].value_counts())
````

<br><br>

<font color=red>**DBSCAN**</font>

<br>

DBSCAN (Density-based spatial clustering of applications with noise) 알고리즘은 반경 내에 있는 <font color=darkblue>**데이터의 밀도를 기준으로 군집을 형성**</font>하는 것입니다.

DBSCAN의 장점은 데이터에 이상치가 존재할 경우 K-Means는 평균 거리 계산에 큰 영향을 주어 효과적인 클러스터링을 수행할 수 없는 반면에 DBSCAN는 알고리즘 과정에서 <font color=darkblue>**어느 군집에도 속하지 못한 데이터는 군집에 포함시키지 않고 이상치로 분류**</font>하는 것입니다.

<br>



````python
val = pd.read_csv('./val.csv')

# 'ID'와 'Class' 변수는 제외합니다.
df = val.drop(columns=['ID', 'Class'])

# 'Class' 변수만 가져옵니다.
label = val[['Class']]
````

````python
# sklearn 라이브러리의 cluster 중에서 DBSCAN를 불러옵니다.
from sklearn.cluster import DBSCAN
````

````python
# 클러스터링을 용이하게 사용하게 별도로 변수로 지정합니다.
# 'eps' 하이퍼파라미터를 사용해 반경의 거리를 설정합니다.
# 'min_samples' 하이퍼파라미터를 사용해 군집에 몇 개의 데이터가 있어야 군집으로 볼 것인지를 설정합니다.
cluster = DBSCAN(eps=0.5, min_samples=30)

array_clusted = cluster.fit_predict(df)

df_clusted = pd.DataFrame(array_clusted)
````

````python
# Pandas 라이브러리의 concat 함수를 활용하여 클렁스터링 결과와 'Class'를 다시 합쳐줍니다.
df_result = pd.concat([df_clusted, label], axis=1)

# 칼럼 명을 변경해줍니다

df_result.columns = ['결과', '정답']
````

````python
# DBSCAN가 다수의 군집과 이상치를 생성하고 구분함을 확인할 수 있습니다.
print(df_result['결과'].value_counts())
print(df_result['정답'].value_counts())
````
