**Ïù¥ÏÉÅÏπòÎ•º Íµ¨Î∂ÑÌïòÎäî Í∏∞Ï§Ä IQR**<br><br>

  * <font color=red>**IQR** (*Interquartile range*)</font>Ïù¥ÎûÄ Ïù¥ÏÉÅÏπòÎ•º Íµ¨Î∂ÑÌï† Ïàò ÏûàÎäî Î∞©Î≤ïÎ°† Ï§ë ÌïòÎÇòÏûÖÎãàÎã§.<br><br>

  * <font color=green>**ÏÇ¨Î∂ÑÏúÑÏàòÎ•º ÌôúÏö©**</font>ÌïòÏó¨ Ïù¥ÏÉÅÏπòÎ•º ÏãùÎ≥ÑÌïòÎäî Í∏∞Ï§ÄÏÑ† Îëê Í∞ú (ÏµúÎåÄÍ∞íÍ≥º ÏµúÏÜåÍ∞í)Î•º ÏÑ§Ï†ïÌï©ÎãàÎã§.<br><br>

  * ÏµúÎåÄÍ∞íÍ≥º ÏµúÏÜåÍ∞íÏùÑ <font color=blue>**Î≤óÏñ¥ÎÇú Îç∞Ïù¥ÌÑ∞Î•º Ïù¥ÏÉÅÏπòÎùºÍ≥† Í∑úÏ†ï**</font>Ìï©ÎãàÎã§.<br><br>

  * ÏµúÎåÄÍ∞íÍ≥º ÏµúÏÜåÍ∞íÏùÄ Îã§ÏùåÏùò ÏàòÏãùÏúºÎ°ú Íµ¨Ìï† Ïàò ÏûàÏäµÎãàÎã§.<br><br>

    - IQR = 3ÏÇ¨Î∂ÑÏúÑÏàò - 1ÏÇ¨Î∂ÄÏúÑÏàò (ÏÉÅÏúÑ 75% ÏßÄÏ†ê Í∞íÍ≥º ÌïòÏúÑ 25% ÏßÄÏ†ê Í∞íÏùò Ï∞®Ïù¥)

    - ÏµúÎåÄÍ∞í = 3ÏÇ¨Î∂ÑÏúÑÏàò + 1.5 * IQR

    - ÏµúÏÜåÍ∞í = 1ÏÇ¨Î∂ÑÏúÑÏàò - 1.5 * IQR

<br>

<br>

**ÏÇ¨Î∂ÑÏúÑÏàò Íµ¨Ìï¥Ï£ºÎäî Ìï®Ïàò <font color=red>quantile()</font>**



```
ÏÉÅÏúÑ 75% ÏßÄÏ†ê Í∞í (3ÏÇ¨Î∂ÑÏúÑÏàò) = Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.75)

ÌïòÏúÑ 25% ÏßÄÏ†ê Í∞í (1ÏÇ¨Î∂ÑÏúÑÏàò) = Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.25)
```

<br>


<br>

**IQR Í∏∞Ï§ÄÏóê Ìï¥ÎãπÌïòÎäî Ïù¥ÏÉÅÏπòÎ•º Îã§Î•∏ Í∞íÏúºÎ°ú Î≥ÄÍ≤ΩÌïòÎäî Î∞©Î≤ï**

```
# IQRÎ•º ÌôúÏö©Ìïú Ïù¥ÏÉÅÏπò Í∏∞Ï§Ä ÏÑ§Ï†ï
IQR = Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.75) - Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.25)
maximum = Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.75) + (1.5 * IQR)
minimum = Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.25) - (1.5 * IQR)

ÌäπÏ†ï Ï°∞Í±¥ 1 = (Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'] > maximum)
ÌäπÏ†ï Ï°∞Í±¥ 2 = (Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'] < minimum)

# Ïù¥ÏÉÅÏπò Í∏∞Ï§ÄÏóê Î∂ÄÌï©Ìïú Îç∞Ïù¥ÌÑ∞Î•º Îã§Î•∏ Í∞íÏúºÎ°ú ÎåÄÏ≤¥

## maximumÏóê Ï¥àÍ≥ºÌïú Îç∞Ïù¥ÌÑ∞Î•º 3ÏÇ¨Î∂ÑÏúÑÍ∞íÏúºÎ°ú Î≥ÄÍ≤Ω
Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ.loc[ÌäπÏ†ï Ï°∞Í±¥ 1] = Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.75)

## minimumÏóê ÎØ∏ÎßåÏù∏ Îç∞Ïù¥ÌÑ∞Î•º 1ÏÇ¨Î∂ÑÏúÑÍ∞íÏúºÎ°ú Î≥ÄÍ≤Ω
Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ.loc[ÌäπÏ†ï Ï°∞Í±¥ 2] = Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.25)
```

<br>

<br>

IQRÎ•º Í∏∞Ï§ÄÏúºÎ°ú Ïù¥ÏÉÅÏπò ÌòπÏùÄ Í∑πÎã®Í∞íÏù¥ Îã§Î•∏ Í∞íÏúºÎ°ú Î≥ÄÍ≤ΩÎêòÏóàÍ∏∞Ïóê 'V1' ÏπºÎüºÏùò **ÌëúÏ§ÄÌé∏Ï∞®Í∞Ä Í∞êÏÜå**ÌïòÏòÄÏùåÏùÑ ÌôïÏù∏Ìï† Ïàò ÏûàÏäµÎãàÎã§.

<br>

````python
# 'V1' ÏπºÎüºÏóê IQRÎ•º Í∏∞Ï§ÄÏúºÎ°ú Ïù¥ÏÉÅÏπò Îç∞Ïù¥ÌÑ∞ Î≥ÄÌôò

print(f'Í∏∞Ï°¥ V1 ÏπºÎüº ÌëúÏ§ÄÌé∏Ï∞® : {train["V1"].std()}\n')

IQR = train['V1'].quantile(0.75) - train['V1'].quantile(0.25)
maximum = train['V1'].quantile(0.75) + (1.5 * IQR)
minimum = train['V1'].quantile(0.25) - (1.5 * IQR)

print(f'IQR : {IQR}')
print(f'maximum : {maximum}')
print(f'minimum : {minimum}\n')

cond1 = (train['V1'] > maximum)
cond2 = (train['V1'] < minimum)

train.loc[cond1] = train['V1'].quantile(0.75)
train.loc[cond2] = train['V1'].quantile(0.25)

print(f'Ïù¥ÏÉÅÏπò Ï≤òÎ¶¨Îêú V1 ÏπºÎüº ÌëúÏ§ÄÌé∏Ï∞® : {train["V1"].std()}')
````

  * <font color=red>**Ïñ∏Îçî ÏÉòÌîåÎßÅ** (*Under Sampling*)</font><br><br>

    - <font color=darkblue>**Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ ÏàòÏóê ÎßûÍ≤å Ï§ÑÏù¥Îäî Î∞©Ïãù**</font>ÏûÖÎãàÎã§.<br><br>

    - Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º Î™®Îëê ÌôúÏö©ÌïòÏßÄ ÏïäÍ∏∞Ïóê Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ÏóêÏÑú <font color=#1589FF>**Ïú†ÏùòÎØ∏Ìïú Ï†ïÎ≥¥Î•º ÏÜêÏã§**</font>Ìï† Ïàò ÏûàÏäµÎãàÎã§.<br><br>


  * <font color=red>**Ïò§Î≤Ñ ÏÉòÌîåÎßÅ** (*Over Sampling*)</font><br><br>
    
    - <font color=darkblue>**ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ ÏàòÏóê ÎßûÍ≤å ÎäòÎ¶¨Îäî Î∞©Ïãù**</font>ÏûÖÎãàÎã§.<br><br>

    - ÎèôÏùºÌïú (ÌòπÏùÄ Ïú†ÏÇ¨Ìïú) ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º ÌôúÏö©ÌïòÍ∏∞Ïóê <font color=#1589FF>**ÏÜåÏàò Î≤îÏ£ºÏùò ÌäπÏ†ï Ï†ïÎ≥¥Í∞Ä Í≥ºÏûâ Î∞òÏòÅ**</font>Îê† Ïàò ÏûàÏäµÎãàÎã§.<br><br>

````python
# Î∂àÍ∑†Ìòï Îç∞Ïù¥ÌÑ∞ Î¨∏Ï†úÎ•º ÌôïÏù∏Ìïú 'Val' Îç∞Ïù¥ÌÑ∞ ÏÖãÏùÑ ÌôúÏö©Ìï©ÎãàÎã§. 

val = pd.read_csv('./val.csv')

print(val['Class'].value_counts())
````

````python
# 'val' Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ Î≥µÏÇ¨ÌïòÏó¨ ÎèôÏùºÌïú Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ ÏÇ¨Ïö©Ìï©ÎãàÎã§.
df = val.copy()

# Î∂àÍ∑†Ìòï Îç∞Ïù¥ÌÑ∞ Î¨∏Ï†úÎ•º Ìï¥Í≤∞Ìï† ÏπºÎüºÏùÑ Íµ¨Î∂ÑÌï©ÎãàÎã§.
# y : Î∂àÍ∑†Ìòï Îç∞Ïù¥ÌÑ∞ Î¨∏Ï†úÍ∞Ä Î∞úÏÉùÌïú ÌäπÏÑ± / X : y Ïù¥Ïô∏Ïùò ÏπºÎüºÎì§ 
y = df[['Class']]
X = df.drop(columns=['Class'])
````

````python
display(X.head(), y.head())
````

<br>

#### Ïñ∏Îçî ÏÉòÌîåÎßÅ

<br>

<font color=red>**Random Under Sampler**</font>



<br>

<font color=darkblue>**Î¨¥ÏûëÏúÑÎ°ú Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º Ï†úÍ±∞**</font>ÌïòÏó¨ Î∂àÍ∑†Ìòï Îç∞Ïù¥ÌÑ∞ Î¨∏Ï†úÎ•º Ìï¥Í≤∞Ìï©ÎãàÎã§.

<br>

````python
# imblearn ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò under_sampling Ï§ëÏóêÏÑú RandomUnderSamplerÎ•º Î∂àÎü¨ÏòµÎãàÎã§.
from imblearn.under_sampling import RandomUnderSampler
````

````python
# ÏÉòÌîåÎßÅÏùÑ Ïö©Ïù¥ÌïòÍ≤å ÏÇ¨Ïö©ÌïòÍ≤å Î≥ÑÎèÑÎ°ú Î≥ÄÏàòÎ°ú ÏßÄÏ†ïÌï©ÎãàÎã§.
under_sampler = RandomUnderSampler()

under_sampled_X, under_sampled_y = under_sampler.fit_resample(X, y)
````

````python
# Pandas ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò concat Ìï®ÏàòÎ•º ÌôúÏö©ÌïòÏó¨ Î∂ÑÎ¶¨Îêú XÏôÄ yÎ•º Îã§Ïãú Ìï©Ï≥êÏ§çÎãàÎã§.
df_random_under = pd.concat([under_sampled_X, under_sampled_y], axis=1)
````

````python
# Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ ÏàòÏôÄ ÎèôÏùºÌïòÍ≤å Î¨¥ÏûëÏúÑÎ°ú Ï∂îÏ∂úÌïòÏòÄÍ∏∞Ïóê Ï¥ù Îç∞Ïù¥ÌÑ∞Ïùò ÏàòÍ∞Ä Í∞êÏÜåÌïòÏòÄÏäµÎãàÎã§.
print(df_random_under.shape)
````

````python
# Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ (Ï†ïÏÉÅ Í±∞Îûò)Ïùò ÏàòÎ•º ÏÜåÏàò Î≤îÏ£ºÏùò Í±∞ÎûòÏùò Îç∞Ïù¥ÌÑ∞ (ÏÇ¨Í∏∞ Í±∞Îûò)Ïùò ÏàòÏôÄ ÎèôÏùºÌïòÍ≤å ÎßûÏ∂îÏñ¥Ï†∏ÏäµÎãàÎã§.
print(df_random_under['Class'].value_counts())
````

Í∑∏Î¶¨Í≥† 'Í∑ºÏ≤òÏóê ÏûàÎã§'ÎùºÎäî ÎßêÏùò Ï†ïÏùòÌïòÎäî 3Í∞ÄÏßÄ Î∞©Î≤ïÏóê Îî∞Îùº Í∞ÅÍ∏∞ Îã§Î•∏ Î∞©ÏãùÏúºÎ°ú Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º Ï∂îÏ∂úÌï©ÎãàÎã§.

NearMissÏóêÏÑú Ï†úÍ≥µÌïòÎäî 'version' ÌïòÏù¥Ìçº ÌååÎùºÎØ∏ÌÑ∞ (Ïö∞Î¶¨Í∞Ä ÏÑ§Ï†ïÌï† Ïàò ÏûàÎäî Í∞í)Ïóê Îî∞Îùº 3Í∞ÄÏßÄ Îã§Î•∏ Under Sampling Î∞©Î≤ïÎ°†ÏùÑ Ï†úÏãúÌï©ÎãàÎã§.

<br>

````python
# imblearn ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò under_sampling Ï§ëÏóêÏÑú NearMissÎ•º Î∂àÎü¨ÏòµÎãàÎã§.
from imblearn.under_sampling import NearMiss
````

<br>

* **Version 1**<br><br>

  - ÌäπÏ†ï Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î°úÎ∂ÄÌÑ∞ Í∞ÄÏû• Í∞ÄÍπåÏö¥ NÍ∞ú (n_neighbors)Ïùò ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ÍπåÏßÄÏùò ÌèâÍ∑† Í±∞Î¶¨Î•º Íµ¨Ìï©ÎãàÎã§.<br><br>

  - Í≥ÑÏÇ∞Îêú ÌèâÍ∑† Í±∞Î¶¨Í∞Ä Í∞ÄÏû• Í∞ÄÍπåÏö¥ MÍ∞úÏùò Îç∞Ïù¥ÌÑ∞Îäî ÎÇ®Í∏∞Í≥† ÎÇòÎ®∏ÏßÄÎäî Ï†úÍ±∞Ìï©ÎãàÎã§.<br><br>

* **Version 2**<br><br>

  - ÌäπÏ†ï Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î°úÎ∂ÄÌÑ∞ Í∞ÄÏû• Î®º Í±∞Î¶¨Ïóê ÏûàÎäî NÍ∞ú (n_neighbors)Ïùò ÏÜåÏàò ÌÅ¥ÎûòÏä§ ÏÉòÌîåÍπåÏßÄÏùò ÌèâÍ∑† Í±∞Î¶¨Î•º Íµ¨Ìï©ÎãàÎã§.<br><br>

  - Í≥ÑÏÇ∞Îêú ÌèâÍ∑† Í±∞Î¶¨Í∞Ä Í∞ÄÏû• Í∞ÄÍπåÏö¥ MÍ∞úÏùò Îç∞Ïù¥ÌÑ∞Îäî ÎÇ®Í∏∞Í≥† ÎÇòÎ®∏ÏßÄÎäî Ï†úÍ±∞Ìï©ÎãàÎã§.<br><br>

* **Version 3**<br><br>

  - Î™®Îì† ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î°úÎ∂ÄÌÑ∞ MÍ∞úÏùò Í∞ÄÏû• Í∞ÄÍπåÏö¥ Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º ÏÑ†ÌÉùÌïúÎã§.<br><br>

  - ÏÑ†ÌÉùÎêú Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ÏóêÏÑúÎ∂ÄÌÑ∞ Í∞ÄÏû• Í∞ÄÍπåÏö¥ NÍ∞ú (n_neighbors)Ïùò ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ÍπåÏßÄÏùò Í±∞Î¶¨Î•º Í≥ÑÏÇ∞ÌïúÎã§.<br><br>

  - ÌèâÍ∑† Í±∞Î¶¨Í∞Ä Í∞ÄÏû• Î®º Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º ÎÇ®Í∏∞Í≥† ÎÇòÎ®∏ÏßÄÎäî Ï†úÍ±∞Ìï©ÎãàÎã§.<br><br>

````python
# ÏÉòÌîåÎßÅÏùÑ Ïö©Ïù¥ÌïòÍ≤å ÏÇ¨Ïö©ÌïòÍ≤å Î≥ÑÎèÑÎ°ú Î≥ÄÏàòÎ°ú ÏßÄÏ†ïÌï©ÎãàÎã§.
# NearMissÏùò Îã§ÏñëÌïú versionÏùÑ ÏÇ¨Ïö©ÌïòÍ≥† Ïã∂ÏúºÎ©¥ Ìï¥Îãπ version Ïô∏Ïùò Í≤ÉÏùÄ Ï£ºÏÑùÏ≤òÎ¶¨ÌïòÎ©¥ Îê©ÎãàÎã§.
under_sampler = NearMiss(version=1, n_neighbors=3)
# under_sampler = NearMiss(version=2, n_neighbors=3)
# under_sampler = NearMiss(version=3, n_neighbors=3)

under_sampled_X, under_sampled_y = under_sampler.fit_resample(X, y)
````

````python
# Pandas ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò concat Ìï®ÏàòÎ•º ÌôúÏö©ÌïòÏó¨ Î∂ÑÎ¶¨Îêú XÏôÄ yÎ•º Îã§Ïãú Ìï©Ï≥êÏ§çÎãàÎã§.
df_near_under = pd.concat([under_sampled_X, under_sampled_y], axis=1)
````

````python
# Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ ÏàòÏôÄ ÎèôÏùºÌïòÍ≤å Î¨¥ÏûëÏúÑÎ°ú Ï∂îÏ∂úÌïòÏòÄÍ∏∞Ïóê Ï¥ù Îç∞Ïù¥ÌÑ∞Ïùò ÏàòÍ∞Ä Í∞êÏÜåÌïòÏòÄÏäµÎãàÎã§.
print(df_near_under.shape)
````

````python
# Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ (Ï†ïÏÉÅ Í±∞Îûò)Ïùò ÏàòÎ•º ÏÜåÏàò Î≤îÏ£ºÏùò Í±∞ÎûòÏùò Îç∞Ïù¥ÌÑ∞ (ÏÇ¨Í∏∞ Í±∞Îûò)Ïùò ÏàòÏôÄ ÎèôÏùºÌïòÍ≤å ÎßûÏ∂îÏñ¥Ï†∏ÏäµÎãàÎã§.
print(df_near_under['Class'].value_counts())
````

<br><br>

<font color=red>**Condensed Nearest Neighbour**</font>

<br>

Condensed Nearest NeighbourÎäî Îã§Ïùå Í≥ºÏ†ïÏùÑ Í±∞Ï≥ê Under SamplingÏùÑ ÏàòÌñâÌï©ÎãàÎã§.<br><br>

  1. ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ Î™®Îëê ÏßëÌï© SÏóê Ìè¨Ìï®ÌïúÎã§.<br><br>

  2. Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ Ï§ëÏóêÏÑú ÌïòÎÇòÎ•º Í≥®ÎùºÏÑú <font color=darkblue>**Í∞ÄÏû• Í∞ÄÍπåÏö¥ Îç∞Ïù¥ÌÑ∞Í∞Ä Îã§Ïàò Î≤îÏ£º Îç∞Ïù¥ÌÑ∞Í∞Ä ÏïÑÎãàÎ©¥ ÏßëÌï© SÏóê Ìè¨Ìï®**</font>ÌïúÎã§.<br><br>

  3. Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Îçî Ïù¥ÏÉÅ ÏÑ†ÌÉùÏßÄÍ∞Ä ÏóÜÏùÑ ÎïåÍπåÏßÄ Í≥ºÏ†ï 2Î•º Î∞òÎ≥µÌïúÎã§.<br><br>

````python
# imblearn ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò under_sampling Ï§ëÏóêÏÑú CondensedNearestNeighbourÎ•º Î∂àÎü¨ÏòµÎãàÎã§.
from imblearn.under_sampling import CondensedNearestNeighbour
````

````python
# ÏÉòÌîåÎßÅÏùÑ Ïö©Ïù¥ÌïòÍ≤å ÏÇ¨Ïö©ÌïòÍ≤å Î≥ÑÎèÑÎ°ú Î≥ÄÏàòÎ°ú ÏßÄÏ†ïÌï©ÎãàÎã§.
under_sampler = CondensedNearestNeighbour()

under_sampled_X, under_sampled_y = under_sampler.fit_resample(X, y)
````

````python
# Pandas ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò concat Ìï®ÏàòÎ•º ÌôúÏö©ÌïòÏó¨ Î∂ÑÎ¶¨Îêú XÏôÄ yÎ•º Îã§Ïãú Ìï©Ï≥êÏ§çÎãàÎã§.
df_cnn_under = pd.concat([under_sampled_X, under_sampled_y], axis=1)
````

````python
# Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ ÏàòÏôÄ ÎèôÏùºÌïòÍ≤å Î¨¥ÏûëÏúÑÎ°ú Ï∂îÏ∂úÌïòÏòÄÍ∏∞Ïóê Ï¥ù Îç∞Ïù¥ÌÑ∞Ïùò ÏàòÍ∞Ä Í∞êÏÜåÌïòÏòÄÏäµÎãàÎã§.
print(df_cnn_under.shape)
````

````python
# ÏïåÍ≥†Î¶¨Ï¶ò ÌäπÏÑ±ÏÉÅ Ï∂îÏ∂úÎêú Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ ÏàòÍ∞Ä ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ ÏàòÏôÄ Í∞ôÏßÄ ÏïäÏùÑ Ïàò ÏûàÏäµÎãàÎã§.
print(df_cnn_under['Class'].value_counts())
````

<font color=red>**Random Over Sampler**</font>

<br>

<font color=darkblue>**Î¨¥ÏûëÏúÑÎ°ú ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º Ï∂îÏ∂ú**</font>ÌïòÏó¨ ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Ïóê Ìè¨Ìï®ÏãúÏºú Î∂àÍ∑†Ìòï Îç∞Ïù¥ÌÑ∞ Î¨∏Ï†úÎ•º Ìï¥Í≤∞Ìï©ÎãàÎã§.

<br>

````python
# imblearn ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò over_sampling Ï§ëÏóêÏÑú RandomOverSamplerÎ•º Î∂àÎü¨ÏòµÎãàÎã§.
from imblearn.over_sampling import RandomOverSampler
````

````python
# ÏÉòÌîåÎßÅÏùÑ Ïö©Ïù¥ÌïòÍ≤å ÏÇ¨Ïö©ÌïòÍ≤å Î≥ÑÎèÑÎ°ú Î≥ÄÏàòÎ°ú ÏßÄÏ†ïÌï©ÎãàÎã§.
over_sampler = RandomOverSampler()

over_sampled_X, over_sampled_y = over_sampler.fit_resample(X, y)
````

````python
# Pandas ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò concat Ìï®ÏàòÎ•º ÌôúÏö©ÌïòÏó¨ Î∂ÑÎ¶¨Îêú XÏôÄ yÎ•º Îã§Ïãú Ìï©Ï≥êÏ§çÎãàÎã§.
df_random_over = pd.concat([over_sampled_X, over_sampled_y], axis=1)
````

````python
# ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ ÏàòÏôÄ ÎèôÏùºÌïòÍ≤å Î¨¥ÏûëÏúÑÎ°ú Ï∂îÏ∂úÌïòÏó¨ Ìè¨Ìï®ÏãúÏº∞Í∏∞Ïóê Ï¥ù Îç∞Ïù¥ÌÑ∞Ïùò ÏàòÍ∞Ä Ï¶ùÍ∞ÄÌïòÏòÄÏäµÎãàÎã§.
print(df_random_over.shape)
````

````python
# ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ (ÏÇ¨Í∏∞ Í±∞Îûò)Ïùò ÏàòÎ•º Îã§Ïàò Î≤îÏ£ºÏùò Í±∞ÎûòÏùò Îç∞Ïù¥ÌÑ∞ (Ï†ïÏÉÅ Í±∞Îûò)Ïùò ÏàòÏôÄ ÎèôÏùºÌïòÍ≤å ÎßûÏ∂îÏñ¥Ï†∏ÏäµÎãàÎã§.
print(df_random_over['Class'].value_counts())
````

````python
# ÎèôÏùºÌïú Îç∞Ïù¥ÌÑ∞Í∞Ä Î™á Í∞úÍ∞Ä ÏûàÎäî ÏßÄ duplicated Ìï®ÏàòÎ•º ÏÇ¨Ïö©ÌïòÏó¨ ÏïåÏïÑÎ¥ÖÏãúÎã§.
# Í∏∞Ï°¥Ïùò Îç∞Ïù¥ÌÑ∞Î•º Í∑∏ÎåÄÎ°ú Í∞ÄÏ†∏ÏôÄ ÏÇ¨Ïö©ÌïòÏòÄÍ∏∞Ïóê Ï§ëÎ≥µÎêú Îç∞Ïù¥ÌÑ∞Í∞Ä ÎßéÏäµÎãàÎã§.
df_same = df_random_over.duplicated()
print(df_same.sum())
````

<br><br>

<font color=red>**SMOTE**</font>

<br>

SMOTE (Synthetic Minority Over-sampling Technique)Îäî Îã§ÏùåÏùò Í≥ºÏ†ïÏùÑ ÏßÑÌñâÌï©ÎãàÎã§.<br><br>

  1. ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ Ï§ëÏóêÏÑú Î¨¥ÏûëÏúÑÎ°ú ÌïòÎÇòÎ•º ÏÑ†ÌÉùÌï©ÎãàÎã§.<br><br>
  
  2. <font color=darkblue>**KNN (K-nearest neighbors) ÏïåÍ≥†Î¶¨Ï¶òÏùÑ ÌôúÏö©**</font>ÌïòÏó¨ Í∞ÄÏû• Í∞ÄÍπåÏö¥ KÍ∞úÏùò ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º Ï∞æÏäµÎãàÎã§.<br><br>

  3. Ï∞æÏùÄ KÍ∞úÏùò ÏÜåÏàò Î≤îÏ£º Îç∞Ïù¥ÌÑ∞ Ï§ë ÌïòÎÇòÎ•º Î¨¥ÏûëÏúÑÎ°ú ÏÑ†ÌÉùÌï©ÎãàÎã§.<br><br>

  4. Ï≤òÏùåÏóê ÏÑ†ÌÉùÌïú Îç∞Ïù¥ÌÑ∞ÏôÄ KNNÎ•º ÌÜµÌï¥ ÏÑ†ÌÉùÎêú Îç∞Ïù¥ÌÑ∞ ÏÇ¨Ïù¥Ïóê ÏÑ†ÏùÑ Í∏ãÏäµÎãàÎã§.<br><br>

  5. <font color=darkblue>**Í∑∏ ÏÑ† ÏúÑÏóê ÏûàÎäî ÏûÑÏùòÏùò Ï†êÏùÑ ÏÑ†ÌÉùÌïòÏó¨ ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞Î•º ÏÉùÏÑ±**</font>Ìï©ÎãàÎã§.<br><br>

  6. ÏõêÌïòÎäî ÎßåÌÅºÏùò Îç∞Ïù¥ÌÑ∞ ÏàòÍ∞Ä ÏÉùÏÑ±Îê† ÎïåÍπåÏßÄ ÏúÑÏùò Í≥ºÏ†ïÏùÑ Î∞òÎ≥µÌï©ÎãàÎã§.<br><br>

````python
# imblearn ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò over_sampling Ï§ëÏóêÏÑú SMOTEÎ•º Î∂àÎü¨ÏòµÎãàÎã§.
from imblearn.over_sampling import SMOTE
````

````python
# ÏÉòÌîåÎßÅÏùÑ Ïö©Ïù¥ÌïòÍ≤å ÏÇ¨Ïö©ÌïòÍ≤å Î≥ÑÎèÑÎ°ú Î≥ÄÏàòÎ°ú ÏßÄÏ†ïÌï©ÎãàÎã§.
over_sampler = SMOTE()

over_sampled_X, over_sampled_y = over_sampler.fit_resample(X, y)
````

````python
# Pandas ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò concat Ìï®ÏàòÎ•º ÌôúÏö©ÌïòÏó¨ Î∂ÑÎ¶¨Îêú XÏôÄ yÎ•º Îã§Ïãú Ìï©Ï≥êÏ§çÎãàÎã§.
df_smote_over = pd.concat([over_sampled_X, over_sampled_y], axis=1)
````

````python
# ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Î•º Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ ÏàòÏôÄ ÎèôÏùºÌïòÍ≤å ÏÉùÏÑ±ÌïòÏòÄÍ∏∞Ïóê Ï¥ù Îç∞Ïù¥ÌÑ∞Ïùò ÏàòÍ∞Ä Ï¶ùÍ∞ÄÌïòÏòÄÏäµÎãàÎã§.
print(df_smote_over.shape)
````

````python
# ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ (ÏÇ¨Í∏∞ Í±∞Îûò)Ïùò ÏàòÎ•º Îã§Ïàò Î≤îÏ£ºÏùò Í±∞ÎûòÏùò Îç∞Ïù¥ÌÑ∞ (Ï†ïÏÉÅ Í±∞Îûò)Ïùò ÏàòÏôÄ ÎèôÏùºÌïòÍ≤å ÎßûÏ∂îÏñ¥Ï†∏ÏäµÎãàÎã§.
print(df_random_over['Class'].value_counts())
````

````python
# ÎèôÏùºÌïú Îç∞Ïù¥ÌÑ∞Í∞Ä Î™á Í∞úÍ∞Ä ÏûàÎäî ÏßÄ duplicated Ìï®ÏàòÎ•º ÏÇ¨Ïö©ÌïòÏó¨ ÏïåÏïÑÎ¥ÖÏãúÎã§.
# Í∏∞Ï°¥ Îç∞Ïù¥ÌÑ∞Î•º Î∞îÌÉïÏúºÎ°ú ÏÉàÎ°≠Í≤å Îç∞Ïù¥ÌÑ∞Î•º ÏÉùÏÑ±ÌïòÏòÄÍ∏∞Ïóê ÏÉÅÎåÄÏ†ÅÏúºÎ°ú Ï§ëÎ≥µÎêú Îç∞Ïù¥ÌÑ∞Í∞Ä Ï†ÅÏäµÎãàÎã§.
df_same = df_random_over.duplicated()
print(df_same.sum())
````

<font color=red>**ADASYN**</font>

<br>

ADASYN (Adaptive Synthetic Sampling)Îäî Í∏∞Ï°¥Ïùò SMOTEÎ•º Î≥¥ÏôÑÌïú Í∏∞Î≤ïÏûÖÎãàÎã§.

SMOTE ÏïåÍ≥†Î¶¨Ï¶òÏóêÏÑú ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ Ï£ºÏúÑÏóê Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Í∞Ä ÏñºÎßàÎÇò ÏûàÎäî ÏßÄ Í≥†Î†§Ìï©ÎãàÎã§.

Í≤∞Î°†Ï†ÅÏúºÎ°ú ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ Ï£ºÎ≥ÄÏóê Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞Í∞Ä ÎßéÏùÑ Í≤ΩÏö∞ Ìï¥Îãπ ÏÜåÏàò Î≤îÏ£º Îç∞Ïù¥ÌÑ∞ÏôÄ Ïú†ÏÇ¨Ìïú Îç∞Ïù¥ÌÑ∞Î•º Îçî ÎßéÏù¥ ÏÉùÏÑ±Ìï©ÎãàÎã§.

Ï¶â, <font color=darkblue>**ÏÜåÏàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ Ï§ëÏóêÏÑú Îã§Ïàò Î≤îÏ£ºÏùò Îç∞Ïù¥ÌÑ∞ÏôÄ Í∞ÄÍπåÏù¥ ÏûàÎäî Îç∞Ïù¥ÌÑ∞Î•º Ï§ëÏã¨ÏúºÎ°ú Ïú†ÏÇ¨Ìïú Îç∞Ïù¥ÌÑ∞Î•º Îçî ÏÉùÏÑ±**</font>Ìï©ÎãàÎã§.

<br>

````python
# imblearn ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò over_sampling Ï§ëÏóêÏÑú ADASYNÎ•º Î∂àÎü¨ÏòµÎãàÎã§.
from imblearn.over_sampling import ADASYN
````

````python
# ÏÉòÌîåÎßÅÏùÑ Ïö©Ïù¥ÌïòÍ≤å ÏÇ¨Ïö©ÌïòÍ≤å Î≥ÑÎèÑÎ°ú Î≥ÄÏàòÎ°ú ÏßÄÏ†ïÌï©ÎãàÎã§.
over_sampler = ADASYN()

over_sampled_X, over_sampled_y = over_sampler.fit_resample(X, y)
````

````python
# Pandas ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò concat Ìï®ÏàòÎ•º ÌôúÏö©ÌïòÏó¨ Î∂ÑÎ¶¨Îêú XÏôÄ yÎ•º Îã§Ïãú Ìï©Ï≥êÏ§çÎãàÎã§.
df_adasyn_over = pd.concat([over_sampled_X, over_sampled_y], axis=1)
````

Ïù¥Ï≤òÎüº <font color=darkblue>**Ïú†ÏÇ¨Ìïú ÏÑ±Í≤© (ÎπÑÏä∑Ìïú Ìå®ÌÑ¥Í≥º Íµ¨Ï°∞)ÏùÑ Í∞ÄÏßÑ Îç∞Ïù¥ÌÑ∞Î•º Í∞ôÏùÄ Íµ∞ÏßëÏúºÎ°ú Î¨∂Ïñ¥ Íµ¨Î∂Ñ**</font>Ìï† Ïàò ÏûàÏäµÎãàÎã§.

Ïù¥Í≤ÉÏùÑ <font color=red>**ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ** (*Clustering*)</font> Ïù¥ÎùºÍ≥† Ìï©ÎãàÎã§.

<br>

ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅÏù¥ ÏûòÎêú Í≤ΩÏö∞, ÌäπÏ†ï Ìïú Íµ∞ÏßëÏóêÎäî Îç∞Ïù¥ÌÑ∞Í∞ÑÏóê ÎÜíÏùÄ Ïú†ÏÇ¨ÏÑ±ÏùÑ Î≥¥Ïù¥Í≥† Íµ∞ÏßëÍ∞ÑÏóêÎäî ÎÇÆÏùÄ Ïú†ÏÇ¨ÏÑ±ÏùÑ Î≥¥Ïó¨Ï§çÎãàÎã§.

<font color=red>**K-Means**</font>

<br>

K-Means ÏïåÍ≥†Î¶¨Ï¶òÏùÄ <font color=darkblue>**Íµ∞ÏßëÏùò Ï§ëÏã¨Í≥º Îç∞Ïù¥ÌÑ∞Í∞ÑÏùò ÌèâÍ∑† Í±∞Î¶¨Î•º ÏÇ¨Ïö©**</font>ÌïòÏó¨ KÍ∞úÏùò Íµ∞ÏßëÏùÑ ÎßåÎì§Ïñ¥ ÎÇ¥Îäî Í≤ÉÏûÖÎãàÎã§.<br><br>

K-Means Îäî Îã§ÏùåÏùò Í≥ºÏ†ïÏùÑ ÏßÑÌñâÌï©ÎãàÎã§.<br><br>


  1. KÍ∞úÏùò ÏûÑÏùò Ï§ëÏã¨Ï†êÏùÑ Î∞∞ÏπòÌï©ÎãàÎã§.<br><br>

  2. Í∞ÅÍ∞ÅÏùò Îç∞Ïù¥ÌÑ∞Îäî Ï§ëÏã¨Ï†êÍ≥ºÏùò ÌèâÍ∑†Í±∞Î¶¨Î•º Í∏∞Ï§ÄÏúºÎ°ú Í∞ÄÏû• Í∞ÄÍπåÏö¥ Ï§ëÏã¨Ï†êÏùÑ Ï§ëÏã¨ÏúºÎ°ú Íµ∞ÏßëÏùÑ ÌòïÏÑ±Ìï©ÎãàÎã§.<br><br>

  3. Ï§ëÏã¨Ï†êÏùÑ Íµ∞ÏßëÏùò Ï§ëÏã¨ÏúºÎ°ú Ïù¥ÎèôÌï©ÎãàÎã§.<br><br>

  4. Í≤∞Í≥ºÍ∞Ä Î∞îÎÄåÏßÄ ÏïäÍ±∞ÎÇò 'max_iter' ÌïòÏù¥Ìçº ÌååÎùºÎØ∏ÌÑ∞Ïóê ÏÑ§Ï†ïÎêú Í∞íÍπåÏßÄ 2Î≤àÍ≥º 3Î≤à Í≥ºÏ†ïÏùÑ Î∞òÎ≥µÌï©ÎãàÎã§.<br><br>

````python
val = pd.read_csv('./val.csv')

# 'ID'ÏôÄ 'Class' Î≥ÄÏàòÎäî Ï†úÏô∏Ìï©ÎãàÎã§.
df = val.drop(columns=['ID', 'Class'])

# 'Class' Î≥ÄÏàòÎßå Í∞ÄÏ†∏ÏòµÎãàÎã§.
label = val[['Class']]
````

````python
# sklearn ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò cluster Ï§ëÏóêÏÑú KMeansÎ•º Î∂àÎü¨ÏòµÎãàÎã§.
from sklearn.cluster import KMeans
````

````python
# ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅÏùÑ Ïö©Ïù¥ÌïòÍ≤å ÏÇ¨Ïö©ÌïòÍ≤å Î≥ÑÎèÑÎ°ú Î≥ÄÏàòÎ°ú ÏßÄÏ†ïÌï©ÎãàÎã§.
# 'n_clusters' ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞Î•º ÏÇ¨Ïö©Ìï¥ ÎßåÎì§Ïñ¥ÎÇº Íµ∞ÏßëÏùò Í∞úÏàòÎ•º ÏÑ§Ï†ïÌï©ÎãàÎã§.
cluster = KMeans(n_clusters=2)

array_clusted = cluster.fit_predict(df)

df_clusted = pd.DataFrame(array_clusted)
````

````python
# Pandas ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò concat Ìï®ÏàòÎ•º ÌôúÏö©ÌïòÏó¨ ÌÅ¥Î†ÅÏä§ÌÑ∞ÎßÅ Í≤∞Í≥ºÏôÄ 'Class'Î•º Îã§Ïãú Ìï©Ï≥êÏ§çÎãàÎã§.
df_result = pd.concat([df_clusted, label], axis=1)

# ÏπºÎüº Î™ÖÏùÑ Î≥ÄÍ≤ΩÌï¥Ï§çÎãàÎã§

df_result.columns = ['Í≤∞Í≥º', 'Ï†ïÎãµ']
````

````python
# ÎåÄÎã§ÏàòÏùò Îç∞Ïù¥ÌÑ∞Îäî Ïûò Íµ¨Î∂ÑÌïòÏòÄÏßÄÎßå Î™á Í∞ú Îç∞Ïù¥ÌÑ∞Îäî ÌãÄÎ¶¥ Í≤ÉÏùÑ ÌôïÏù∏Ìï† Ïàò ÏûàÏäµÎãàÎã§.
print(df_result['Í≤∞Í≥º'].value_counts())
print(df_result['Ï†ïÎãµ'].value_counts())
````

<font color=red>**Mean-Shift Clustering**</font>

<br>

Mean-Shift Clustering ÏïåÍ≥†Î¶¨Ï¶òÏùÄ <font color=darkblue>**Îç∞Ïù¥ÌÑ∞Í∞Ä Î™®Ïó¨ÏûàÎäî Î∞ÄÎèÑÎ•º Í∏∞Ï§ÄÏúºÎ°ú Íµ∞ÏßëÏùÑ ÌòïÏÑ±**</font>Ìï©ÎãàÎã§.

K-MeansÏôÄ ÎòêÎã§Î•∏ Ï∞®Ïù¥Ï†êÏù¥ ÏûàÎã§Î©¥ K-MeansÎäî Ï¥àÍ∏∞Ïóê Íµ∞ÏßëÏùò Í∞úÏàòÎ•º ÏÑ§Ï†ïÌï¥Ï£ºÏñ¥Ïïº ÌïòÏßÄÎßå Mean-ShiftÎäî ÏÑ§Ï†ïÌï¥Ï§Ä ÎåÄÏó≠Ìè≠Ïùò Î≤îÏúÑÏóê Îî∞Îùº ÏûêÎèôÏ†ÅÏúºÎ°ú Íµ∞ÏßëÏùò Í∞úÏàòÍ∞Ä Ï†ïÌï¥ÏßëÎãàÎã§.

<br>

````python
val = pd.read_csv('./val.csv')

# 'ID'ÏôÄ 'Class' Î≥ÄÏàòÎäî Ï†úÏô∏Ìï©ÎãàÎã§.
df = val.drop(columns=['ID', 'Class'])

# 'Class' Î≥ÄÏàòÎßå Í∞ÄÏ†∏ÏòµÎãàÎã§.
label = val[['Class']]
````

````python
# sklearn ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò cluster Ï§ëÏóêÏÑú MeanShiftÎ•º Î∂àÎü¨ÏòµÎãàÎã§.
from sklearn.cluster import MeanShift
````

````python
# ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅÏùÑ Ïö©Ïù¥ÌïòÍ≤å ÏÇ¨Ïö©ÌïòÍ≤å Î≥ÑÎèÑÎ°ú Î≥ÄÏàòÎ°ú ÏßÄÏ†ïÌï©ÎãàÎã§.
# 'bandwidth' ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞Î•º ÏÇ¨Ïö©Ìï¥ ÎåÄÏó≠Ìè≠Ïùò Î≤îÏúÑÎ•º ÏÑ§Ï†ïÌï©ÎãàÎã§.
cluster = MeanShift(bandwidth=0.1)

array_clusted = cluster.fit_predict(df)

df_clusted = pd.DataFrame(array_clusted)
````

````python
# Pandas ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò concat Ìï®ÏàòÎ•º ÌôúÏö©ÌïòÏó¨ ÌÅ¥Î†ÅÏä§ÌÑ∞ÎßÅ Í≤∞Í≥ºÏôÄ 'Class'Î•º Îã§Ïãú Ìï©Ï≥êÏ§çÎãàÎã§.
df_result = pd.concat([df_clusted, label], axis=1)

# ÏπºÎüº Î™ÖÏùÑ Î≥ÄÍ≤ΩÌï¥Ï§çÎãàÎã§

df_result.columns = ['Í≤∞Í≥º', 'Ï†ïÎãµ']
````

````python
# Mean-ShiftÍ∞Ä ÎåÄÏó≠Ìè≠Ïóê Îî∞Îùº 2Í∞úÏùò Íµ∞ÏßëÏù¥ ÏïÑÎãå Îã§ÏàòÏùò Íµ∞ÏßëÏùÑ ÏÉùÏÑ±Ìï®ÏùÑ ÌôïÏù∏Ìï† Ïàò ÏûàÏäµÎãàÎã§.
print(df_result['Í≤∞Í≥º'].value_counts())
print(df_result['Ï†ïÎãµ'].value_counts())
````

<font color=red>**DBSCAN**</font>

<br>

DBSCAN (Density-based spatial clustering of applications with noise) ÏïåÍ≥†Î¶¨Ï¶òÏùÄ Î∞òÍ≤Ω ÎÇ¥Ïóê ÏûàÎäî <font color=darkblue>**Îç∞Ïù¥ÌÑ∞Ïùò Î∞ÄÎèÑÎ•º Í∏∞Ï§ÄÏúºÎ°ú Íµ∞ÏßëÏùÑ ÌòïÏÑ±**</font>ÌïòÎäî Í≤ÉÏûÖÎãàÎã§.

DBSCANÏùò Ïû•Ï†êÏùÄ Îç∞Ïù¥ÌÑ∞Ïóê Ïù¥ÏÉÅÏπòÍ∞Ä Ï°¥Ïû¨Ìï† Í≤ΩÏö∞ K-MeansÎäî ÌèâÍ∑† Í±∞Î¶¨ Í≥ÑÏÇ∞Ïóê ÌÅ∞ ÏòÅÌñ•ÏùÑ Ï£ºÏñ¥ Ìö®Í≥ºÏ†ÅÏù∏ ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅÏùÑ ÏàòÌñâÌï† Ïàò ÏóÜÎäî Î∞òÎ©¥Ïóê DBSCANÎäî ÏïåÍ≥†Î¶¨Ï¶ò Í≥ºÏ†ïÏóêÏÑú <font color=darkblue>**Ïñ¥Îäê Íµ∞ÏßëÏóêÎèÑ ÏÜçÌïòÏßÄ Î™ªÌïú Îç∞Ïù¥ÌÑ∞Îäî Íµ∞ÏßëÏóê Ìè¨Ìï®ÏãúÌÇ§ÏßÄ ÏïäÍ≥† Ïù¥ÏÉÅÏπòÎ°ú Î∂ÑÎ•ò**</font>ÌïòÎäî Í≤ÉÏûÖÎãàÎã§.

<br>



````python
val = pd.read_csv('./val.csv')

# 'ID'ÏôÄ 'Class' Î≥ÄÏàòÎäî Ï†úÏô∏Ìï©ÎãàÎã§.
df = val.drop(columns=['ID', 'Class'])

# 'Class' Î≥ÄÏàòÎßå Í∞ÄÏ†∏ÏòµÎãàÎã§.
label = val[['Class']]
````

````python
# sklearn ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò cluster Ï§ëÏóêÏÑú DBSCANÎ•º Î∂àÎü¨ÏòµÎãàÎã§.
from sklearn.cluster import DBSCAN
````

````python
# ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅÏùÑ Ïö©Ïù¥ÌïòÍ≤å ÏÇ¨Ïö©ÌïòÍ≤å Î≥ÑÎèÑÎ°ú Î≥ÄÏàòÎ°ú ÏßÄÏ†ïÌï©ÎãàÎã§.
# 'eps' ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞Î•º ÏÇ¨Ïö©Ìï¥ Î∞òÍ≤ΩÏùò Í±∞Î¶¨Î•º ÏÑ§Ï†ïÌï©ÎãàÎã§.
# 'min_samples' ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞Î•º ÏÇ¨Ïö©Ìï¥ Íµ∞ÏßëÏóê Î™á Í∞úÏùò Îç∞Ïù¥ÌÑ∞Í∞Ä ÏûàÏñ¥Ïïº Íµ∞ÏßëÏúºÎ°ú Î≥º Í≤ÉÏù∏ÏßÄÎ•º ÏÑ§Ï†ïÌï©ÎãàÎã§.
cluster = DBSCAN(eps=0.5, min_samples=30)

array_clusted = cluster.fit_predict(df)

df_clusted = pd.DataFrame(array_clusted)
````

````python
# Pandas ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò concat Ìï®ÏàòÎ•º ÌôúÏö©ÌïòÏó¨ ÌÅ¥Î†ÅÏä§ÌÑ∞ÎßÅ Í≤∞Í≥ºÏôÄ 'Class'Î•º Îã§Ïãú Ìï©Ï≥êÏ§çÎãàÎã§.
df_result = pd.concat([df_clusted, label], axis=1)

# ÏπºÎüº Î™ÖÏùÑ Î≥ÄÍ≤ΩÌï¥Ï§çÎãàÎã§

df_result.columns = ['Í≤∞Í≥º', 'Ï†ïÎãµ']
````

````python
# DBSCANÍ∞Ä Îã§ÏàòÏùò Íµ∞ÏßëÍ≥º Ïù¥ÏÉÅÏπòÎ•º ÏÉùÏÑ±ÌïòÍ≥† Íµ¨Î∂ÑÌï®ÏùÑ ÌôïÏù∏Ìï† Ïàò ÏûàÏäµÎãàÎã§.
print(df_result['Í≤∞Í≥º'].value_counts())
print(df_result['Ï†ïÎãµ'].value_counts())
````
