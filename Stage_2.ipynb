## 3 - 1 전처리와 특성 공학은 무엇인가요?

<br>

**전처리 (Preprocessing)**

  * 우리가 시험을 준비하는 모습을 떠올려 보십다. 교재에 오타가 있으면 수정하기도 합니다. 그리고 교재에서 설명하는 내용이 어려우면 이해하기 쉬운 형태로 바꾸기도 합니다.

  * **데이터 분석 과정에서 더 좋은 인사이트를 발견하기 위해서** 혹은 **머신러닝 학습 과정에서 더 나은 성능을 내기 위해**, <font color=red>**기존의 데이터를 잘 정돈되게 정리**</font>하여 위의 과정에 적합한 형태의 데이터로 가공하는 것입니다.

<br>

**특성 공학 (Feature Engineering)**

  * 다시 한 번 더 시험을 준비하는 모습을 떠올려 보십다. 공부하는 과목의 핵심 개념을 이해하기위해 좋은 정리 노트를 찾기도 하고 만들기도 합니다. 그리고 시험 직전에는 시험 출제 빈도가 높은 개념만 따로 적은 노트만 보는 친구들의 모습을 보기도 합니다.

  * 특징 공학은 전처리와 마찬가지로 머신러닝 학습 과정에서 더 나은 성능을 내기 위해, <font color=red>**데이터의 특징이 잘 드러내도록**</font> 새로운 특징을 만들어내거나 불필요한 특징은 제거하는 방법 등으로 데이터를 가공하는 것을 뜻합니다.

<br>

**전처리와 특성 공학의 필요성**

  * Stage 1에서 배웠 듯이 데이터 수집 과정이 완벽하지 않아 데이터 값이 존재하지 않은 결측치가 발생할 수 있습니다. 결측치가 존재할 경우, 결측치를 제거하거나 다른 값으로 채워야 합니다.

  * 기존의 데이터가 우리에게 인사이트를 주기에는 불완전할 수 있습니다. 예를 들어 보면 '축구를 좋아하는 고객의 남녀 각각의 특징 조사'에서 여성 고객에 관하여 인사이트를 얻기에는 데이터 수가 부족할 수 도 있습니다.

  * 예외적인 상황을 고려할 필요가 있습니다. 예를 들어 '치킨에 대하여 소비액에 조사하였을 때, 월드컵 기간의 치킨 소비량이 일반적인 치킨 소비량'이라고 보기는 힘듭니다.

  * 도메인 지식을 통해 좋은 데이터 특징을 발굴할 수 도 있습니다. 신용카드 결재에 미치는 중요한 요소로 불경기 여부, 축제 혹은 명절 여부 등이 있습니다.

  * 반대로 불필요한 데이터 특징은 제거할 수 도 있습니다. MZ세대는 너무 광범위한 나이대를 포함합니다. 그래서 MZ세대 여부를 물어보지 않는 것이 고객의 인사이트를 찾아내기 더 쉬울 수 있습니다.

원하는 조건에 다양할 수 있습니다. 

각각의 경우를 살펴봅시다.

**특정 조건 1에 부합하고 특정 조건 2에도 부합한 데이터 (교집합)**



```
특정 조건 1 = (데이터프레임['특정 칼럼 1'] >= 특정 값)
특정 조건 2 = (데이터프레임['특정 칼럼 2'] <= 특정 값)

데이터프레임.loc[(특정 조건 1 & 특정 조건 2)]
```



````python
# V1 칼럼에서 -0.1보다 크면서 V2에서 0.1보다 작은 데이터만 추출
cond1 = (train['V1'] >= -0.1)
cond2 = (train['V2'] <= 0.1)

display(train.loc[cond1 & cond2])
````

**특정 조건 1에 부합하기도 하고 특정 조건 2에도 부합하기도 한 데이터 (합집합)**



```
특정 조건 1 = (데이터프레임['특정 칼럼 1'] >= 특정 값)
특정 조건 2 = (데이터프레임['특정 칼럼 2'] <= 특정 값)

데이터프레임.loc[(특정 조건 1 | 특정 조건 2)]
```



````python
# V1 칼럼에서 -0.1보다 크거나 V2에서 0.1보다 작은 데이터만 추출
cond1 = (train['V1'] >= -0.1)
cond2 = (train['V2'] <= 0.1)

display(train.loc[cond1 | cond2])
````

**특정 조건에 부합하지 않은 데이터 (여집합)**



```
특정 조건 = (데이터프레임['특정 칼럼 1'] >= 특정 값)

데이터프레임.loc[(~특정 조건)]
```



````python
# 'V1' 칼럼에서 0 이하가 아닌 데이터만 추출

cond = (train['V1'] <= 0)

display(train.loc[~cond])
````

### 결측치 처리

<br>

결측치를 처리할 수 있는 방법 크게 결측치가 있는 데이터를 삭제하는 제거 (Delection)와 결측치를 다른 값으로 채우는 대치 (Imputation)가 있습니다.

우리가 사용하는 데이터 셋에는 결측치가 존재하지 않습니다.

그래서 인위적으로 결측치를 만들어 보겠습니다.

````python
# random 라이브러리를 활용하여 무작위 수를 생성할 수 있습니다.
# randrange는 무작위 정수 값을 생성할 수 있습니다.
from random import randrange

# numpy 라이버러리를 활용하여 array 데이터를 다룰 수 있습니다.
# NaN는 결측값을 생성합니다.
from numpy import NaN
````

````python
# 'train' 데이터에서 'V1' 칼럼만 사용합니다
df = train[['V1']]

# 무작위로 'df' 데이터의 값을 결측값으로 변경합니다
list_null = []
for _ in range(100):

  index_null = randrange(df.shape[0])
  list_null.append(index_null)

list_null = sorted(list_null)

df.iloc[list_null] = NaN
````

````python
# 결측치로 변경된 데이터의 인덱스
print(list_null[:5])
````

````python
# Stage 1를 기억하며 결측값이 몇 개 존재하는 지 확인해봅시다
df.isnull().sum()
````

### 테이블 핸들링을 활용한 이상치 변환

**이상치를 구분하는 기준 IQR**

  * IQR (Interquartile range)이란 이상치를 구분할 수 있는 방법론 중 하나입니다.

  * 사분위수를 활용화여 이상치를 식별하는 기준선 두 개 (최대값과 최소값)를 설정합니다.

  * 최대값과 최소값을 벗어난 데이터를 이상치라고 규정합니다.

  * 최대값과 최소값은 다음의 수식으로 구할 수 있습니다.

    - IQR = 3사분위수 - 1사부위수 (상위 75% 지점 값과 하위 25% 지점 값의 차이)

    - 최대값 = 3사분위수 + 1.5 * IQR

    - 최소값 = 1사분위수 - 1.5 * IQR

**사분위수 구해주는 함수 quantile()**



```
상위 75% 지점 값 (3사분위수) = 데이터프레임['특정 칼럼'].quantile(0.75)

하위 25% 지점 값 (1사분위수) = 데이터프레임['특정 칼럼'].quantile(0.25)
```



**IQR 기준에 해당하는 이상치를 다른 값으로 변경하는 방법**

```
# IQR를 활용한 이상치 기준 설정
IQR = 데이터프레임['특정 칼럼'].quantile(0.75) - 데이터프레임['특정 칼럼'].quantile(0.25)
maximum = 데이터프레임['특정 칼럼'].quantile(0.75) + (1.5 * IQR)

## minimum에 미만인 데이터를 1사분위값으로 변경

```



IQR를 기준으로 이상치 혹은 극단값이 다른 값으로 변경되었기에 'V1' 칼럼의 표준편차가 감소하였음을 확인할 수 있습니다.

````python
# 'V1' 칼럼에 IQR를 기준으로 이상치 데이터 변환

print(f'기존 V1 칼럼 표준편차 : {train["V1"].std()}\n')

IQR = train['V1'].quantile(0.75) - train['V1'].quantile(0.25)
maximum = train['V1'].quantile(0.75) + (1.5 * IQR)
minimum = train['V1'].quantile(0.25) - (1.5 * IQR)

print(f'IQR : {IQR}')
print(f'maximum : {maximum}')
print(f'minimum : {minimum}\n')

cond1 = (train['V1'] > maximum)
cond2 = (train['V1'] < minimum)

train.loc[cond1] = train['V1'].quantile(0.75)
train.loc[cond2] = train['V1'].quantile(0.25)

print(f'이상치 처리된 V1 칼럼 표준편차 : {train["V1"].std()}')
````

### 스케일링

<br>

여러분들은 IQR를 활용하여 데이터의 Feature의 범위를 조정시켜 보았습니다.

이러한 과정을 데이터 스케일링 (Data Scaling)이라고 합니다.

데이터 스케일링에는 다양한 기법들이 존재합니다.

하나씩 알아가보도록 합시다.

**Scikit-Learn (sklearn)**

* sklearn 라이브러리는 데이터 전처리, 특성공학 그리고 머신러닝 모델 생성 등 머신러닝 엔지니어링 기술을 활용할 수 있도록 도와주는 친구입니다. 

사이킷런.png

````python
# 'train' 데이터 셋에서 'ID' 칼럼은 제외하고 사용합니다.
df = train.drop(columns=['ID'])
````

#### Standard Scaler

Standard Scaler는 데이터의 Feature의 평균을 0 그리고 분산을 1로 스케일링하여 데이터 표준 정규 분포를 따르도록 합니다.

스크린샷 2023-01-27 14.28.22.png

````python
# sklearn 라이브러리에서 preprocessing 중에서 StandardScaler를 불러옵니다.
from sklearn.preprocessing import StandardScaler
````

````python
# 스케일링을 용이하게 사용하게 별도로 변수로 지정합니다.
stand_scaler = StandardScaler()

# fit_transform 함수를 활용해 StandardScaler를 적용시킨 후 array 형태로 반환합니다.
array_stand_scaled = stand_scaler.fit_transform(df)

# array 형태의 데이터를 pandas 라이브러리의 DataFrame 함수를 사용해 데이터프레임로 변환합니다.
df_stand_scaled = pd.DataFrame(array_stand_scaled)

display(df_stand_scaled.head())
````

````python
# 평균과 표준편차가 어떻게 변하였는 지 확인해봅시다.
# 각각의 Feature들이 평균은 0에 가깝게 표준편차는 1에 가깝게 변화하였음을 볼 수 있습니다.
display(df.describe(), df_stand_scaled.describe())
````

#### MinMax Scaler

MinMaxScaler는 데이터의 Feature의 최소값을 0 그리고 최대값을 1로 스케일링하여 데이터가 0과 1사이에 분포하도록 합니다.

스크린샷 2023-01-27 14.28.45.png

````python
# sklearn 라이브러리에서 preprocessing 중에서 MinMaxScaler를 불러옵니다.
from sklearn.preprocessing import MinMaxScaler
````

````python
# 스케일링을 용이하게 사용하게 별도로 변수로 지정합니다.
minmax_scaler = MinMaxScaler()

# fit_transform 함수를 활용해 StandardScaler를 적용시킨 후 array 형태로 반환합니다.
array_minmax_scaled = minmax_scaler.fit_transform(df)

# array 형태의 데이터를 pandas 라이브러리의 DataFrame 함수를 사용해 데이터프레임로 변환합니다.
df_minmax_scaled = pd.DataFrame(array_minmax_scaled)

display(df_minmax_scaled.head())
````

````python
# 최소값과 최대값이 어떻게 변하였는 지 확인해봅시다.
# 각각의 Feature들이 최소값은 0 그리고 최대값은 1로 변화하였음을 볼 수 있습니다.
display(df.describe(), df_minmax_scaled.describe())
````

#### Robust Scaler

RobustScaler는 데이터의 Feature의 중앙값을 0 그리고 IQR이 1로 스케일링합니다.

````python
# sklearn 라이브러리에서 preprocessing 중에서 RobustScaler를 불러옵니다.
from sklearn.preprocessing import RobustScaler
````

````python
# 스케일링을 용이하게 사용하게 별도로 변수로 지정합니다.
robust_scaler = RobustScaler()

# fit_transform 함수를 활용해 StandardScaler를 적용시킨 후 array 형태로 반환합니다.
array_robust_scaled = robust_scaler.fit_transform(df)

# array 형태의 데이터를 pandas 라이브러리의 DataFrame 함수를 사용해 데이터프레임로 변환합니다.
df_robust_scaled = pd.DataFrame(array_robust_scaled)

display(df_robust_scaled.head())
````

````python
# 1사분위수와 3사분위수가 어떻게 변하였는 지 확인해봅시다.
# 각각의 Feature들이 1사분위수와 3사분위수를 확인하여 IQR를 구해보면 1에 가깝게 변환하였음을 볼 수 있습니다.
display(df.describe(), df_robust_scaled.describe())
````

#### 불귤형 데이터이란?

Stage 1에서 신용카드 거래 중에서 사기거래가 발생할 경우는 극히 드물다고 하였습니다.

그렇기에 수집된 데이터에 신용카드 거래 데이터 (val.csv) 중에서 사기거래인 경우는 30개에 불과했습니다.

데이터의 관측치 분포가 균등하지 못하고 극심히 차이가 나는 걸 불균형 데이터 (Imbalanced Data)라고 합니다.

**불균형 데이터가 줄 수 있는 문제점**

  * 또다시 시험을 예로 들어봅시다. 국어, 영어, 수학 과목 중 국어 과목과 관련된 교재와 문제집은 많이 공부하고 나머지는 적게 공부한다면 국어 시험에서는 좋은 성적을 볼 수 있어도 영어와 수학에서 좋은 성적을 기대하기는 어렵습니다. 

  * 데이터 수가 많은 관측치에서만 인사이트를 얻을 수 있으며, 데이터 수가 적은 데이터는 충분히 활용하지 못할 수 있습니다.

  * 머신러닝이 데이터가 적은 관측치는 충분히 학습하지 못하여 수가 적은 관측치 대해서는 충분한 성능을 발휘하지 못할 수 있습니다.

#### 언더 샘플링 VS 오버 샘플링

불균형 데이터 문제를 샘플링을 통해 해결할 수 있습니다.

샘플링이란 불균형한 데이터 분포를 균등하게 맞추기 위해 특정 데이터를 추출해가는 방식입니다.

샘플링 기법에는 크게 언더 샘플링과 오버 샘플링이 있습니다.

  * 언더 샘플링 (Under Sampling) 은 다수 범주의 데이터를 소수 범주의 데이터 수에 맞게 줄이는 방식입니다.

  * 오버 샘플링 (Over Sampling) 은 소수의 범주의 데이터를 다수 범주의 데이터 수에 맞게 늘리는 방식입니다.
