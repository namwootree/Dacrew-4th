### ÌÖåÏù¥Î∏î Ìï∏Îì§ÎßÅÏùÑ ÌôúÏö©Ìïú Ïù¥ÏÉÅÏπò Î≥ÄÌôò

**Ïù¥ÏÉÅÏπòÎ•º Íµ¨Î∂ÑÌïòÎäî Í∏∞Ï§Ä IQR**

  * IQR (Interquartile range)Ïù¥ÎûÄ Ïù¥ÏÉÅÏπòÎ•º Íµ¨Î∂ÑÌï† Ïàò ÏûàÎäî Î∞©Î≤ïÎ°† Ï§ë ÌïòÎÇòÏûÖÎãàÎã§.

  * ÏÇ¨Î∂ÑÏúÑÏàòÎ•º ÌôúÏö©ÌôîÏó¨ Ïù¥ÏÉÅÏπòÎ•º ÏãùÎ≥ÑÌïòÎäî Í∏∞Ï§ÄÏÑ† Îëê Í∞ú (ÏµúÎåÄÍ∞íÍ≥º ÏµúÏÜåÍ∞í)Î•º ÏÑ§Ï†ïÌï©ÎãàÎã§.

  * ÏµúÎåÄÍ∞íÍ≥º ÏµúÏÜåÍ∞íÏùÑ Î≤óÏñ¥ÎÇú Îç∞Ïù¥ÌÑ∞Î•º Ïù¥ÏÉÅÏπòÎùºÍ≥† Í∑úÏ†ïÌï©ÎãàÎã§.

  * ÏµúÎåÄÍ∞íÍ≥º ÏµúÏÜåÍ∞íÏùÄ Îã§ÏùåÏùò ÏàòÏãùÏúºÎ°ú Íµ¨Ìï† Ïàò ÏûàÏäµÎãàÎã§.

    - IQR = 3ÏÇ¨Î∂ÑÏúÑÏàò - 1ÏÇ¨Î∂ÄÏúÑÏàò (ÏÉÅÏúÑ 75% ÏßÄÏ†ê Í∞íÍ≥º ÌïòÏúÑ 25% ÏßÄÏ†ê Í∞íÏùò Ï∞®Ïù¥)

    - ÏµúÎåÄÍ∞í = 3ÏÇ¨Î∂ÑÏúÑÏàò + 1.5 * IQR

    - ÏµúÏÜåÍ∞í = 1ÏÇ¨Î∂ÑÏúÑÏàò - 1.5 * IQR

**ÏÇ¨Î∂ÑÏúÑÏàò Íµ¨Ìï¥Ï£ºÎäî Ìï®Ïàò quantile()**



```
ÏÉÅÏúÑ 75% ÏßÄÏ†ê Í∞í (3ÏÇ¨Î∂ÑÏúÑÏàò) = Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.75)

ÌïòÏúÑ 25% ÏßÄÏ†ê Í∞í (1ÏÇ¨Î∂ÑÏúÑÏàò) = Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.25)
```



**IQR Í∏∞Ï§ÄÏóê Ìï¥ÎãπÌïòÎäî Ïù¥ÏÉÅÏπòÎ•º Îã§Î•∏ Í∞íÏúºÎ°ú Î≥ÄÍ≤ΩÌïòÎäî Î∞©Î≤ï**

```
# IQRÎ•º ÌôúÏö©Ìïú Ïù¥ÏÉÅÏπò Í∏∞Ï§Ä ÏÑ§Ï†ï
IQR = Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.75) - Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.25)
maximum = Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.75) + (1.5 * IQR)
minimum = Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.25) - (1.5 * IQR)

ÌäπÏ†ï Ï°∞Í±¥ 1 = (Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'] > maximum)
ÌäπÏ†ï Ï°∞Í±¥ 2 = (Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'] < minimum)

# Ïù¥ÏÉÅÏπò Í∏∞Ï§ÄÏóê Î∂ÄÌï©Ìïú Îç∞Ïù¥ÌÑ∞Î•º Îã§Î•∏ Í∞íÏúºÎ°ú ÎåÄÏ≤¥

## maximumÏóê Ï¥àÍ≥ºÌïú Îç∞Ïù¥ÌÑ∞Î•º 3ÏÇ¨Î∂ÑÏúÑÍ∞íÏúºÎ°ú Î≥ÄÍ≤Ω
Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ.loc[ÌäπÏ†ï Ï°∞Í±¥ 1] = Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.75)

## minimumÏóê ÎØ∏ÎßåÏù∏ Îç∞Ïù¥ÌÑ∞Î•º 1ÏÇ¨Î∂ÑÏúÑÍ∞íÏúºÎ°ú Î≥ÄÍ≤Ω
Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ.loc[ÌäπÏ†ï Ï°∞Í±¥ 2] = Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ['ÌäπÏ†ï ÏπºÎüº'].quantile(0.25)
```



IQRÎ•º Í∏∞Ï§ÄÏúºÎ°ú Ïù¥ÏÉÅÏπò ÌòπÏùÄ Í∑πÎã®Í∞íÏù¥ Îã§Î•∏ Í∞íÏúºÎ°ú Î≥ÄÍ≤ΩÎêòÏóàÍ∏∞Ïóê 'V1' ÏπºÎüºÏùò ÌëúÏ§ÄÌé∏Ï∞®Í∞Ä Í∞êÏÜåÌïòÏòÄÏùåÏùÑ ÌôïÏù∏Ìï† Ïàò ÏûàÏäµÎãàÎã§.

````python
# 'V1' ÏπºÎüºÏóê IQRÎ•º Í∏∞Ï§ÄÏúºÎ°ú Ïù¥ÏÉÅÏπò Îç∞Ïù¥ÌÑ∞ Î≥ÄÌôò

print(f'Í∏∞Ï°¥ V1 ÏπºÎüº ÌëúÏ§ÄÌé∏Ï∞® : {train["V1"].std()}\n')

IQR = train['V1'].quantile(0.75) - train['V1'].quantile(0.25)
maximum = train['V1'].quantile(0.75) + (1.5 * IQR)
minimum = train['V1'].quantile(0.25) - (1.5 * IQR)

print(f'IQR : {IQR}')
print(f'maximum : {maximum}')
print(f'minimum : {minimum}\n')

cond1 = (train['V1'] > maximum)
cond2 = (train['V1'] < minimum)

train.loc[cond1] = train['V1'].quantile(0.75)
train.loc[cond2] = train['V1'].quantile(0.25)

print(f'Ïù¥ÏÉÅÏπò Ï≤òÎ¶¨Îêú V1 ÏπºÎüº ÌëúÏ§ÄÌé∏Ï∞® : {train["V1"].std()}')
````
