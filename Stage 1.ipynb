<br>**지도 학습**

* 풀어보아야 하는 문제 (Feature : 설명 변수)와 문제에 대한 정답 (Label : 목표 변수)를 컴퓨터에 주어 학습시키고 틀리는 문제가 적어지도록 학습합니다.

**비지도 학습**

* 데이터 사이언티스로 성장하는 길에 정답이 받으시 존재한다면 얼마나 편할지 모릅니다.

## 3 - 6 데이터가 다양하다구요?

**데이터의 유형**

* 데이터는 크게 언어와 문자와 같은 정성적 데이터와 수치, 도형, 기호와 같은 정량적 데이터로 구분됩니다.

**데이터의 속성**

* 데이터의 속성은 범주형 데이터와 수치형 데이터로 속성을 구분할 수 있습니다.<br><br>
* 범주형에는 명목형과 순서형이 있습니다
  - 명목형 : 값이 순서와 상관없이 의미만 구분되는 경우 (ex : 성별, 지역, 부서)
  - 순서형 : 값이 기준에 따라 순서를 의미하는 경우 (ex : 학년, 등급, 날짜)<br><br>
* 수치형에는 이산형과 연속형이 있습니다
  - 이산형 : 값을 하나하나 셀 수 있는 경우 (ex : 사기 거래 개수, 시간 당 결제 횟수)
  - 연속형 : 값을 하나하나 셀 수 없는 경우 (ex : 키, 온도)
